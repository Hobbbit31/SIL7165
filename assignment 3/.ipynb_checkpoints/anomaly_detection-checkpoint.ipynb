# Advanced Anomaly Detection

This notebook detects various types of anomalous behaviour in network traffic data from `../b/network_analysis_data*.csv`.

**Sections:**
- (a) Statistical Traffic Analysis — thresholds for packet size, flow counts, protocol distribution
- (b) Behavioural Analysis — sudden behavioural changes, coordinated source activity
- (c) Suspicious Communication Patterns — long-duration flows, multi-protocol IPs

```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (13, 5)

csv_files = sorted(glob.glob('b/*.csv'))

df = pd.concat([pd.read_csv(f, low_memory=False) for f in csv_files], ignore_index=True)

# Parse datetimes and numeric cols
df['startDateTime'] = pd.to_datetime(df['startDateTime'], errors='coerce')
df['stopDateTime']  = pd.to_datetime(df['stopDateTime'],  errors='coerce')
for col in ['totalSourceBytes', 'totalDestinationBytes',
            'totalSourcePackets', 'totalDestinationPackets']:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

df['totalBytes'] = df['totalSourceBytes'] + df['totalDestinationBytes']
df['hour']       = df['startDateTime'].dt.floor('H')
df['day']        = df['startDateTime'].dt.floor('D')

print(f'Loaded {len(df):,} flows')
df.head(3)
```

## (a) Statistical Traffic Analysis

### Statistical Thresholds

We define anomaly thresholds for three traffic metrics:

1. **Packet Sizes** — The distribution is highly right-skewed (many small flows, few huge ones). Z-score is poor for skewed data; we use **IQR method** (Q3 + 1.5×IQR) to flag large-flow anomalies.
2. **Flow Counts** — Hourly flow counts follow a roughly normal distribution; we use **mean + 2×std** to flag short-term bursts.
3. **Protocol Distribution** — We establish a baseline % per protocol. Sudden shifts in this distribution suggest unusual activity (e.g., a new scanning protocol appearing).

```python
# ── Packet size distribution ─────────────────────────────────────────────────
fig, ax = plt.subplots()
pos = df[df['totalBytes'] > 0]['totalBytes']
ax.hist(np.log10(pos + 1), bins=100, color='steelblue', edgecolor='none')
ax.set_title('Packet Size Distribution (log10 scale)')
ax.set_xlabel('log10(totalBytes + 1)')
ax.set_ylabel('Flow Count')
plt.tight_layout(); plt.show()

# ── Hourly flow count ────────────────────────────────────────────────────────
flows_per_hour = df.groupby('hour').size()
fig, ax = plt.subplots()
flows_per_hour.plot(ax=ax, lw=0.8, color='darkorange')
ax.set_title('Hourly Flow Counts')
ax.set_xlabel('Time'); ax.set_ylabel('Number of Flows')
plt.tight_layout(); plt.show()

# ── Protocol distribution baseline ──────────────────────────────────────────
protocol_pct = df['protocolName'].value_counts(normalize=True) * 100
print('Baseline Protocol Distribution (%):')
print(protocol_pct.round(2).to_string())
```

```python
# ── Compute thresholds ────────────────────────────────────────────────────────
# Packet size threshold (IQR)
q1_pkt = df['totalBytes'].quantile(0.25)
q3_pkt = df['totalBytes'].quantile(0.75)
iqr_pkt = q3_pkt - q1_pkt
pkt_threshold = q3_pkt + 1.5 * iqr_pkt

# Hourly flow count threshold (mean + 2*std)
flow_mean = flows_per_hour.mean()
flow_std  = flows_per_hour.std()
flow_upper = flow_mean + 2 * flow_std

# Daily flow count threshold (mean + 1*std)
flows_per_day = df.groupby('day').size()
daily_upper = flows_per_day.mean() + flows_per_day.std()

print(f'Packet size threshold (IQR)  : {pkt_threshold:.2f} bytes')
print(f'Hourly flow threshold         : {flow_upper:.2f} flows/hr')
print(f'Daily flow threshold          : {daily_upper:.2f} flows/day')
```

```python
# ── Temporal comparison: hourly vs daily ─────────────────────────────────────
fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=False)

flows_per_hour.plot(ax=axes[0], lw=0.8, color='tab:blue')
axes[0].axhline(flow_upper, color='red', linestyle='--', label=f'Threshold = {flow_upper:.0f}')
axes[0].set_title('Hourly Flow Counts')
axes[0].set_ylabel('Flows/hr')
axes[0].legend()

flows_per_day.plot(ax=axes[1], lw=1.2, color='tab:orange')
axes[1].axhline(daily_upper, color='red', linestyle='--', label=f'Threshold = {daily_upper:.0f}')
axes[1].set_title('Daily Flow Totals')
axes[1].set_ylabel('Flows/day')
axes[1].legend()

plt.tight_layout(); plt.show()

# Classify anomalous days
anomalous_days = flows_per_day[flows_per_day > daily_upper]
for day in anomalous_days.index:
    hourly = flows_per_hour[flows_per_hour.index.date == day.date()]
    if hourly.empty:
        continue
    max_h = hourly.max()
    label = 'Short-term burst (single spike)' if max_h > flow_upper else 'Sustained elevated traffic'
    print(f'  {day.date()}  daily={anomalous_days[day]:,}  max_hourly={max_h:,}  → {label}')
```

```python
# ── IPs generating unusually high or low traffic ──────────────────────────────
src_traffic = df.groupby('source')['totalSourceBytes'].sum()

q1_s = src_traffic.quantile(0.25)
q3_s = src_traffic.quantile(0.75)
iqr_s = q3_s - q1_s

low_ips  = src_traffic[src_traffic < q1_s - 1.5 * iqr_s]
high_ips = src_traffic[src_traffic > q3_s + 1.5 * iqr_s]

print(f'Low-traffic outlier IPs  : {len(low_ips)}')
print(f'High-traffic outlier IPs : {len(high_ips)}')
print('\nTop high-traffic outliers:')
print(high_ips.sort_values(ascending=False).head(10).to_string())
```

## (b) Behavioural Analysis

### Sudden Behavioural Changes
We identify source IPs that were inactive for at least one hour and then suddenly sent > 50,000 bytes in the next hour. This pattern may indicate:
- A dormant bot waking up
- A compromised host being activated
- A scheduled malicious job

```python
SPIKE_THRESHOLD = 50000  # bytes per hour

traffic_by_hour = (
    df.groupby(['source', 'hour'])['totalSourceBytes']
      .sum()
      .unstack(fill_value=0)
)
traffic_diffs = traffic_by_hour.diff(axis=1).fillna(0)

behavioral_spikes = []
for ip in traffic_by_hour.index:
    series = traffic_by_hour.loc[ip]
    diffs  = traffic_diffs.loc[ip]
    idle   = (series == 0)
    idle_prev = idle.shift(1).fillna(True)
    if any(idle_prev & (diffs > SPIKE_THRESHOLD)):
        behavioral_spikes.append(ip)

print(f'IPs with sudden spikes (>{SPIKE_THRESHOLD/1000:.0f} KB) after inactivity: {len(behavioral_spikes)}')
for ip in behavioral_spikes[:30]:
    print(f'  {ip}')
```

### Network-Wide Correlation Patterns

We detect destination IPs contacted by an unusually high number of **distinct source IPs** within a single hour — a hallmark of coordinated or distributed attack activity (e.g., DDoS, worm propagation).

**Threshold**: Q3 + 1.5×IQR over all (destination, hour) unique source counts. If IQR = 0 (flat distribution), fallback to 99.5th percentile.

```python
contact_counts = (
    df.groupby(['destination', 'hour'])['source']
      .nunique()
      .unstack(fill_value=0)
)
counts_flat = contact_counts.stack()

q1_cc = counts_flat.quantile(0.25)
q3_cc = counts_flat.quantile(0.75)
iqr_cc = q3_cc - q1_cc
threshold_cc = (q3_cc + 1.5 * iqr_cc) if iqr_cc > 0 else counts_flat.quantile(0.995)

anomalies_cc = counts_flat[counts_flat > threshold_cc]
print(f'Threshold : {threshold_cc:.1f} unique sources/hr')
print(f'Anomalous (destination, hour) pairs: {len(anomalies_cc)}')
print('\nTop anomalies (destination IPs contacted by most unique sources in a single hour):')
print(anomalies_cc.sort_values(ascending=False).head(15).to_string())
```

## (c) Suspicious Communication Patterns

### Long-Duration Connections

We flag flows that last significantly longer than typical — these may indicate:
- Persistent C2 channels
- Misconfigured services keeping sockets open
- Data exfiltration via slow, steady connections

**Threshold**: Q3 + 1.5×IQR on flow duration (seconds). Fallback to 99.5th percentile if IQR = 0.

```python
df['duration'] = (df['stopDateTime'] - df['startDateTime']).dt.total_seconds()

# Distribution
pos_dur = df[df['duration'] > 0]['duration']
fig, ax = plt.subplots()
ax.hist(np.log10(pos_dur + 1), bins=100, color='mediumpurple', edgecolor='none')
ax.set_title('Connection Duration Distribution (log10 scale)')
ax.set_xlabel('log10(duration_seconds + 1)')
ax.set_ylabel('Flow Count')
plt.tight_layout(); plt.show()

q1_d = df['duration'].quantile(0.25)
q3_d = df['duration'].quantile(0.75)
iqr_d = q3_d - q1_d
dur_threshold = (q3_d + 1.5 * iqr_d) if iqr_d > 0 else df['duration'].quantile(0.995)

long_flows = df[df['duration'] > dur_threshold]
print(f'Duration threshold       : {dur_threshold:.2f} seconds')
print(f'Long-duration flow count : {len(long_flows):,}')
print('\nTop 10 longest flows:')
print(
    long_flows.sort_values('duration', ascending=False)
    [['source', 'destination', 'protocolName', 'duration']]
    .head(10)
    .to_string(index=False)
)
```

### IPs Using Multiple Protocols in Short Time Windows

An IP using many different protocols within a 15-minute window is suspicious — it may be conducting:
- Multi-vector scanning
- Protocol tunnelling / evasion
- Automated reconnaissance

**Time window**: 15 minutes (captures short bursts; hourly would mask brief multi-protocol flurries).

```python
df['window_15m'] = df['startDateTime'].dt.floor('15min')

proto_counts = df.groupby(['source', 'window_15m'])['protocolName'].nunique()

q1_p = proto_counts.quantile(0.25)
q3_p = proto_counts.quantile(0.75)
iqr_p = q3_p - q1_p
proto_threshold = (q3_p + 1.5 * iqr_p) if iqr_p > 0 else proto_counts.quantile(0.995)

suspicious_proto = proto_counts[proto_counts > proto_threshold]
print(f'Protocol diversity threshold: {proto_threshold:.1f} protocols per 15-min window')
print(f'Suspicious entries: {len(suspicious_proto)}')
print('\nSuspicious IPs communicating using multiple protocols in a short time (15 min window):')
print(suspicious_proto.sort_values(ascending=False).head(20).to_string())
```
