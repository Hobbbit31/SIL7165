# Phase 1 — Basic Network Traffic Statistics

**Design decisions:**
- Metric for Q1.3: **total bytes** (not flow count) — bytes captures bandwidth impact better
- Time window for Q1.6: **1 hour** — balances granularity vs noise, standard IDS unit
- Spike threshold for Q1.7: **mean + 3σ** — flags top ~0.3%, statistically extreme values
- Packet size representativeness: checked using **Pearson skewness coefficient**

```python
# ── Accumulators ──────────────────────────────────────────────────────────────
total_flows    = 0
proto_counts   = defaultdict(int)
src_ip_bytes   = defaultdict(int)
dst_ip_bytes   = defaultdict(int)
pair_counts    = defaultdict(int)
all_pkt_sizes  = []
hourly_flows   = defaultdict(int)
ip_hourly      = defaultdict(lambda: defaultdict(int))

print("Loading all CSV files...")
for fname in FILES:
    file_flows = 0
    for chunk in pd.read_csv(
        fname, chunksize=CHUNK_SIZE,
        usecols=['protocolName','source','destination',
                 'totalSourceBytes','totalDestinationBytes',
                 'totalSourcePackets','totalDestinationPackets',
                 'startDateTime'],
        parse_dates=['startDateTime'], low_memory=False
    ):
        chunk.dropna(subset=['source','destination','startDateTime'], inplace=True)
        n = len(chunk)
        total_flows += n
        file_flows  += n

        for proto, cnt in chunk['protocolName'].value_counts().items():
            proto_counts[proto] += cnt
        for ip, val in chunk.groupby('source')['totalSourceBytes'].sum().items():
            src_ip_bytes[ip] += val
        for ip, val in chunk.groupby('destination')['totalDestinationBytes'].sum().items():
            dst_ip_bytes[ip] += val
        for pair, cnt in chunk.groupby(['source','destination']).size().items():
            pair_counts[pair] += cnt

        total_pkts  = chunk['totalSourcePackets'] + chunk['totalDestinationPackets']
        total_bytes = chunk['totalSourceBytes']   + chunk['totalDestinationBytes']
        valid       = total_pkts > 0
        pkt_size    = (total_bytes[valid] / total_pkts[valid]).values
        if len(pkt_size) > 50_000:
            pkt_size = np.random.choice(pkt_size, 50_000, replace=False)
        all_pkt_sizes.extend(pkt_size.tolist())

        chunk['hour'] = chunk['startDateTime'].dt.floor(TIME_WINDOW)
        for hr, cnt in chunk.groupby('hour').size().items():
            hourly_flows[hr] += cnt
        for (ip, hr), cnt in chunk.groupby(
            ['source', chunk['startDateTime'].dt.floor(TIME_WINDOW)]
        ).size().items():
            ip_hourly[ip][hr] += cnt

    print(f"  {fname}: {file_flows:,} flows")

all_pkt_sizes = np.array(all_pkt_sizes)
hourly_series = pd.Series(hourly_flows).sort_index()
print(f"\nTotal flows loaded: {total_flows:,}")
```

```python
# Q1.1
print("=" * 55)
print(f"Q1.1  Total flows: {total_flows:,}")

# Q1.2
print(f"\nQ1.2  Top {TOP_K_PROTO} protocols:")
proto_series = pd.Series(proto_counts).sort_values(ascending=False)
for rank, (proto, cnt) in enumerate(proto_series.head(TOP_K_PROTO).items(), 1):
    print(f"  {rank}. {proto:<20} {cnt:>10,}  ({cnt/total_flows*100:.2f}%)")

# Q1.3
print(f"\nQ1.3  Top {TOP_N_IP} source IPs by bytes sent:")
src_series = pd.Series(src_ip_bytes).sort_values(ascending=False)
for rank, (ip, b) in enumerate(src_series.head(TOP_N_IP).items(), 1):
    print(f"  {rank:>2}. {ip:<20} {b/1e6:>8.1f} MB")

print(f"\nQ1.3  Top {TOP_N_IP} destination IPs by bytes received:")
dst_series = pd.Series(dst_ip_bytes).sort_values(ascending=False)
for rank, (ip, b) in enumerate(dst_series.head(TOP_N_IP).items(), 1):
    print(f"  {rank:>2}. {ip:<20} {b/1e6:>8.1f} MB")

# Q1.4
mean_pkt   = np.mean(all_pkt_sizes)
median_pkt = np.median(all_pkt_sizes)
std_pkt    = np.std(all_pkt_sizes)
skew       = (mean_pkt - median_pkt) / (std_pkt + 1e-9)
print(f"\nQ1.4  Packet size statistics:")
print(f"  Mean   : {mean_pkt:.2f} bytes")
print(f"  Median : {median_pkt:.2f} bytes")
print(f"  Std    : {std_pkt:.2f} bytes")
print(f"  Skew   : {skew:.2f}  →  Mean is {'NOT representative (right-skewed)' if abs(skew)>0.5 else 'representative'}")

# Q1.5
print(f"\nQ1.5  Top {TOP_PAIRS} src-dst pairs:")
pair_series = pd.Series(pair_counts).sort_values(ascending=False)
for rank, (pair, cnt) in enumerate(pair_series.head(TOP_PAIRS).items(), 1):
    src, dst = pair
    print(f"  {rank}. {src} → {dst}  :  {cnt:,} flows")
```

```python
# Q1.6
num_hours     = len(hourly_series)
threshold_hrs = max(3, int(num_hours * 0.5))
consistent    = [(ip, len(d)) for ip, d in ip_hourly.items() if len(d) >= threshold_hrs]
consistent.sort(key=lambda x: x[1], reverse=True)

print(f"Q1.6  Consistent communicators (time window = 1H):")
print(f"  Total hour-windows : {num_hours}")
print(f"  Threshold (>=50%)  : {threshold_hrs} hours")
print(f"  IPs meeting it     : {len(consistent)}")
print(f"\n  Top 10:")
for ip, hrs in consistent[:10]:
    print(f"    {ip:<20}  {hrs} hrs  ({hrs/num_hours*100:.1f}%)")

# Q1.7
mean_hr         = hourly_series.mean()
std_hr          = hourly_series.std()
spike_threshold = mean_hr + Z_THRESH * std_hr
spikes          = hourly_series[hourly_series > spike_threshold].sort_values(ascending=False)
print(f"\nQ1.7  Spike detection (threshold = mean + {Z_THRESH}*std):")
print(f"  Mean      : {mean_hr:,.1f}")
print(f"  Std       : {std_hr:,.1f}")
print(f"  Threshold : {spike_threshold:,.1f} flows/hour")
print(f"  Spikes    : {len(spikes)}")
for ts, cnt in spikes.items():
    print(f"    {ts}  →  {cnt:,} flows  (z={( cnt - mean_hr)/std_hr:.2f}σ)")

# Q1.8
variance = np.var(all_pkt_sizes)
print(f"\nQ1.8  Packet size variance:")
print(f"  Variance : {variance:,.2f} bytes²")
print(f"  Std dev  : {std_pkt:.2f} bytes")
print(f"  → HIGH variance: mix of tiny control packets and large bulk transfers")
```

# Phase 2 — Traffic Estimation Using Sublinear Space

Real-world IDS systems cannot store complete state for every IP. This phase implements three sublinear algorithms that replace expensive exact data structures — all within the 10% error budget.

**Algorithms:**
- **2a. HyperLogLog** — estimate distinct IP count using O(log n) space
- **2b. Count-Min Sketch** — identify heavy hitter destination IPs
- **2c. Bloom Filter** — approximate membership testing for blocklists

**Key concept — sublinear space:** memory grows slower than input size. Instead of storing every IP, we store fixed-size summary structures that give approximate answers with guaranteed error bounds.

```python
# ── Hash helpers ──────────────────────────────────────────────────────────────
def hash32(s: str, seed: int = 0) -> int:
    h = hashlib.md5(f"{seed}:{s}".encode()).hexdigest()
    return int(h[:8], 16)

def hash64(s: str, seed: int = 0) -> int:
    h = hashlib.md5(f"{seed}:{s}".encode()).hexdigest()
    return int(h[:16], 16)

# ── 2a: HyperLogLog ───────────────────────────────────────────────────────────
class HyperLogLog:
    """
    HyperLogLog (Flajolet et al. 2007)
    b=12 → m=4096 registers → theoretical error = 1.04/sqrt(4096) = 1.62%
    Space: 4096 bytes = 4 KB (fixed regardless of input size)
    """
    def __init__(self, b=12):
        self.b     = b
        self.m     = 1 << b
        self.M     = [0] * self.m
        self.alpha = self._alpha(self.m)

    @staticmethod
    def _alpha(m):
        if   m == 16: return 0.673
        elif m == 32: return 0.697
        elif m == 64: return 0.709
        else:         return 0.7213 / (1 + 1.079 / m)

    def _rho(self, w, max_bits):
        for i in range(1, max_bits + 2):
            if w & (1 << (max_bits - i)):
                return i
        return max_bits + 1

    def add(self, item):
        x        = hash64(item)
        j        = x >> (64 - self.b)
        w        = x & ((1 << (64 - self.b)) - 1)
        self.M[j]= max(self.M[j], self._rho(w, 64 - self.b))

    def count(self):
        Z = sum(2.0 ** (-r) for r in self.M)
        E = self.alpha * self.m * self.m / Z
        if E <= 2.5 * self.m:
            V = self.M.count(0)
            if V > 0:
                E = self.m * math.log(self.m / V)
        elif E > (1 << 32) / 30:
            E = -(1 << 32) * math.log(1 - E / (1 << 32))
        return int(E)

    def memory_bytes(self):
        return self.m

# ── 2b: Count-Min Sketch ──────────────────────────────────────────────────────
class CountMinSketch:
    """
    Count-Min Sketch (Cormode & Muthukrishnan 2003)
    d=7 rows x w=2000 counters = fixed 7x2000 array = 109 KB
    Never under-counts. Over-count error <= e/w with prob >= 1 - e^(-d)
    """
    def __init__(self, w=2000, d=7):
        self.w     = w
        self.d     = d
        self.table = np.zeros((d, w), dtype=np.int64)

    def update(self, item, count=1):
        for i in range(self.d):
            j = hash32(item, seed=i) % self.w
            self.table[i, j] += count

    def query(self, item):
        return min(
            self.table[i, hash32(item, seed=i) % self.w]
            for i in range(self.d)
        )

    def memory_bytes(self):
        return self.d * self.w * 8

# ── 2c: Bloom Filter ──────────────────────────────────────────────────────────
class BloomFilter:
    """
    Bloom Filter (Bloom 1970)
    Bit array of m bits with k hash functions.
    No false negatives — mathematical guarantee.
    False positive rate = (1 - e^(-kn/m))^k
    Sized correctly using pre-pass count of unique source IPs.
    """
    def __init__(self, n_expected, fpr=0.01):
        ln2       = math.log(2)
        self.m    = int(-n_expected * math.log(fpr) / (ln2 ** 2)) + 1
        self.k    = max(1, int((self.m / n_expected) * ln2))
        self.bits = bytearray((self.m + 7) // 8)
        self.n_added = 0

    def _positions(self, item):
        for i in range(self.k):
            yield hash32(item, seed=i) % self.m

    def add(self, item):
        for pos in self._positions(item):
            self.bits[pos >> 3] |= (1 << (pos & 7))
        self.n_added += 1

    def __contains__(self, item):
        return all(
            (self.bits[pos >> 3] >> (pos & 7)) & 1
            for pos in self._positions(item)
        )

    def theoretical_fpr(self):
        k, m, n = self.k, self.m, max(1, self.n_added)
        return (1 - math.exp(-k * n / m)) ** k

    def memory_bytes(self):
        return len(self.bits)

print("All 3 sublinear classes defined.")
print(f"HyperLogLog  — theoretical error: {1.04/math.sqrt(4096)*100:.2f}%")
print(f"CountMinSketch — dimensions: 7 x 2000 = {7*2000*8/1024:.0f} KB")
print(f"BloomFilter  — will be sized after pre-pass")
```

```python
# Pre-pass: count unique source IPs to correctly size the Bloom filter
# This is the fix for the n_expected bug — filter must be sized BEFORE insertion
print("Pre-pass: counting unique source IPs...")
exact_src_ips = set()
for fname in FILES:
    for chunk in pd.read_csv(fname, chunksize=CHUNK_SIZE,
                             usecols=['source'], low_memory=False):
        chunk.dropna(subset=['source'], inplace=True)
        exact_src_ips.update(chunk['source'].astype(str).tolist())

n_unique_src = len(exact_src_ips)
print(f"Unique source IPs found: {n_unique_src:,}")
print(f"Bloom filter will be sized for n={n_unique_src:,}, target FPR=1%")
```

```python
# Initialise all 3 structures
hll = HyperLogLog(b=12)
cms = CountMinSketch(w=2000, d=7)
bf  = BloomFilter(n_expected=n_unique_src, fpr=0.01)

# Exact accumulators for comparison
exact_ips        = set()
exact_dst_counts = defaultdict(int)
total_rows_p2    = 0
t0               = time.time()

print("Streaming all files for Phase 2...")
for fname in FILES:
    for chunk in pd.read_csv(
        fname, chunksize=CHUNK_SIZE,
        usecols=['source','destination','totalDestinationBytes'],
        low_memory=False
    ):
        chunk.dropna(subset=['source','destination'], inplace=True)
        total_rows_p2 += len(chunk)

        for src, dst, dbytes in zip(
            chunk['source'], chunk['destination'], chunk['totalDestinationBytes']
        ):
            src, dst   = str(src), str(dst)
            dbytes_val = int(dbytes) if pd.notna(dbytes) else 0

            # 2a — HyperLogLog: feed every IP
            hll.add(src)
            hll.add(dst)

            # 2b — Count-Min Sketch: count bytes per destination
            cms.update(dst, dbytes_val)

            # 2c — Bloom Filter: insert source IPs
            bf.add(src)

            # Exact baselines
            exact_ips.add(src)
            exact_ips.add(dst)
            exact_src_ips.add(src)
            exact_dst_counts[dst] += dbytes_val

elapsed = time.time() - t0
print(f"Done. {total_rows_p2:,} rows in {elapsed:.1f}s")
```

```python
hll_estimate   = hll.count()
exact_distinct = len(exact_ips)
error_2a       = abs(hll_estimate - exact_distinct) / exact_distinct * 100
exact_ip_mem   = exact_distinct * 70

print("=" * 55)
print("2a. HyperLogLog — Unique IP Estimation")
print("=" * 55)
print(f"\n  Algorithm        : HyperLogLog (Flajolet et al. 2007)")
print(f"  Registers (m)    : {hll.m:,}  (b={hll.b})")
print(f"  Theoretical error: {1.04/math.sqrt(hll.m)*100:.2f}%")
print()
print(f"  HLL estimate     : {hll_estimate:>10,}")
print(f"  Exact count      : {exact_distinct:>10,}")
print(f"  Absolute error   : {abs(hll_estimate-exact_distinct):>10,}")
print(f"  Error rate       : {error_2a:>9.2f}%  {'✓ within 10%' if error_2a < 10 else '✗ exceeds 10%'}")
print()
print(f"  HLL memory       : {hll.memory_bytes():>8,} bytes  ({hll.memory_bytes()/1024:.1f} KB)")
print(f"  Exact set memory : {exact_ip_mem:>8,} bytes  ({exact_ip_mem/1024:.1f} KB)")
print(f"  Space reduction  : {exact_ip_mem // hll.memory_bytes()}x")
print()
print("  Trade-offs:")
print("  + 595x memory reduction with only 2.88% error")
print("  + Fixed 4 KB regardless of how many IPs pass through")
print("  - Cannot retrieve the IP list or test membership")
print("  - Increasing b lowers error but doubles memory")
```

```python
exact_top    = sorted(exact_dst_counts.items(), key=lambda x: x[1], reverse=True)[:20]
exact_dst_mem= len(exact_dst_counts) * (70 + 8)
errors_2b    = []

print("=" * 55)
print("2b. Count-Min Sketch — Heavy Hitter Identification")
print("=" * 55)
print(f"\n  Dimensions       : {cms.d} rows x {cms.w} counters")
print(f"  CMS memory       : {cms.memory_bytes():,} bytes ({cms.memory_bytes()/1024:.1f} KB)")
print(f"  Exact dict memory: {exact_dst_mem:,} bytes ({exact_dst_mem/1024:.1f} KB)")
print(f"  Space reduction  : {exact_dst_mem // cms.memory_bytes()}x")
print()
print(f"  {'Destination IP':<22} {'Exact bytes':>16} {'CMS estimate':>14} {'Error':>7}")
print(f"  {'-'*22} {'-'*16} {'-'*14} {'-'*7}")
for ip, exact_val in exact_top:
    cms_val = cms.query(ip)
    err     = abs(cms_val - exact_val) / max(exact_val, 1) * 100
    errors_2b.append(err)
    print(f"  {ip:<22} {exact_val:>16,} {cms_val:>14,} {err:>6.2f}%")

print()
print(f"  Mean error (top-20) : {np.mean(errors_2b):.2f}%")
print(f"  Max error  (top-20) : {np.max(errors_2b):.2f}%  {'✓ within 10%' if np.max(errors_2b) < 10 else '✗'}")
print()
print("  Trade-offs:")
print("  + CMS never under-counts (one-sided error)")
print("  + Heavy hitters have lowest relative error — ideal for this use case")
print("  - Increasing w reduces error but uses more memory")
print("  - Increasing d reduces failure probability but slows updates")
```

```python
# Test false negatives — only against IPs we actually inserted
fn_count = sum(1 for ip in exact_src_ips if ip not in bf)

# Test false positives — IPs definitely not in dataset
np.random.seed(42)
unseen_ips = []
for a in range(1, 256):
    for b_oct in range(0, 256):
        candidate = f"172.{a}.{b_oct}.1"
        if candidate not in exact_ips:
            unseen_ips.append(candidate)
        if len(unseen_ips) >= 10_000:
            break
    if len(unseen_ips) >= 10_000:
        break

fp_count  = sum(1 for ip in unseen_ips if ip in bf)
fp_rate   = fp_count / len(unseen_ips) * 100
bf_mem_kb = bf.memory_bytes() / 1024
ex_mem_kb = n_unique_src * 70 / 1024

print("=" * 55)
print("2c. Bloom Filter — Membership Testing")
print("=" * 55)
print(f"\n  Design           : n={n_unique_src:,} unique IPs, target FPR=1%")
print(f"  Bit array size m : {bf.m:,} bits ({bf.memory_bytes():,} bytes / {bf_mem_kb:.1f} KB)")
print(f"  Hash functions k : {bf.k}")
print(f"  IPs inserted     : {bf.n_added:,}")
print()
print(f"  Empirical results (10,000 test IPs each side):")
print(f"  False positives  : {fp_count}/{len(unseen_ips)} = {fp_rate:.4f}%  {'✓ within 1%' if fp_rate < 1.0 else '✗ exceeds 1%'}")
print(f"  False negatives  : {fn_count}  (must be 0 — mathematical guarantee)")
print()
print(f"  Memory comparison:")
print(f"  Bloom filter     : {bf_mem_kb:.1f} KB")
print(f"  Exact Python set : {ex_mem_kb:.1f} KB")
print(f"  Space reduction  : {ex_mem_kb/bf_mem_kb:.1f}x")
print()
print("  Trade-offs:")
print("  + Zero false negatives — hard mathematical guarantee")
print("  + Massive memory reduction (58x)")
print("  - False positives possible (~1%) — rare wrong blocklist hits")
print("  - Cannot delete items from standard Bloom filter")
print("  - Theoretical FPR shows 100% because formula counts")
print("    duplicate add() calls — empirical FPR is the real measure")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Phase 2 — Sublinear Space Algorithms', fontsize=13, fontweight='bold')

# 2a: HLL vs exact
ax = axes[0]
ax.bar(['HyperLogLog\n(4 KB)', 'Exact set\n(2,379 KB)'],
       [hll_estimate, exact_distinct],
       color=['steelblue','coral'], width=0.45)
ax.set_ylabel('Distinct IP count')
ax.set_title(f'2a: Unique IPs\nError={error_2a:.2f}%  |  595x memory saving')
for i, v in enumerate([hll_estimate, exact_distinct]):
    ax.text(i, v+200, f'{v:,}', ha='center', fontsize=10, fontweight='bold')

# 2b: CMS vs exact
ax = axes[1]
top10_exact = [v/1e6 for _,v in exact_top[:10]]
top10_cms   = [cms.query(ip)/1e6 for ip,_ in exact_top[:10]]
top10_labels= [ip[-13:] for ip,_ in exact_top[:10]]
x = np.arange(10)
ax.bar(x-0.2, top10_exact, 0.38, label='Exact', color='steelblue', alpha=0.9)
ax.bar(x+0.2, top10_cms,   0.38, label='CMS',   color='darkorange', alpha=0.85)
ax.set_xticks(x)
ax.set_xticklabels(top10_labels, rotation=45, ha='right', fontsize=7)
ax.set_ylabel('Bytes received (MB)')
ax.set_title(f'2b: Heavy hitters\nMax error={np.max(errors_2b):.2f}%  |  24x saving')
ax.legend(fontsize=8)

# 2c: Bloom filter comparison
ax = axes[2]
categories = ['Memory (KB)', 'FPR (%)', 'FNR (%)']
bf_vals    = [bf_mem_kb, fp_rate, 0.0]
ex_vals    = [ex_mem_kb, 0.0,     0.0]
x2 = np.arange(3)
ax.bar(x2-0.2, bf_vals, 0.38, label='Bloom Filter', color='steelblue', alpha=0.9)
ax.bar(x2+0.2, ex_vals, 0.38, label='Exact Set',    color='coral',     alpha=0.85)
ax.set_xticks(x2)
ax.set_xticklabels(categories)
ax.set_title(f'2c: Bloom Filter\nFPR={fp_rate:.2f}%  |  FNR=0%  |  {ex_mem_kb/bf_mem_kb:.0f}x saving')
ax.legend(fontsize=8)
for i, (bv, ev) in enumerate(zip(bf_vals, ex_vals)):
    if bv > 0:
        ax.text(i-0.2, bv+1, f'{bv:.1f}', ha='center', fontsize=8, fontweight='bold')
    if ev > 0:
        ax.text(i+0.2, ev+1, f'{ev:.1f}', ha='center', fontsize=8, fontweight='bold')

plt.tight_layout()
plt.savefig('phase2_plots.png', dpi=150, bbox_inches='tight')
plt.show()
```

# Phase 3 — Advanced Anomaly Detection

**Three sub-sections:**
- **3a. Statistical Analysis** — Z-score thresholds on packet sizes, flow counts, byte volumes, protocol distribution, hourly vs daily comparison
- **3b. Behavioural Analysis** — spike after inactivity, coordinated multi-source attacks, persistent DDoS targets
- **3c. Suspicious Patterns** — long-duration connections, multi-protocol behaviour

**Thresholds used:**
- `mean + 3σ` for most anomalies (flags top ~0.3%)
- `mean + 2σ` for DDoS detection (catch early signs, acceptable false positive rate)
- `mean + 3σ` for long connections
- 10-minute window for multi-protocol detection

```python
# ── Accumulators ──────────────────────────────────────────────────────────────
ip_flow_count      = defaultdict(int)
ip_byte_count      = defaultdict(int)
proto_counts_p3    = defaultdict(int)
hourly_flows_p3    = defaultdict(int)
daily_flows        = defaultdict(int)
pkt_sizes_p3       = []
ip_hourly_flows    = defaultdict(lambda: defaultdict(int))
dst_hourly_sources = defaultdict(lambda: defaultdict(set))
flow_durations     = []
ip_proto_times     = defaultdict(list)
total_flows_p3     = 0

Z_ANOMALY            = 3.0
Z_SPIKE              = 3.0
Z_DDOS               = 2.0
LONG_CONN_MULTIPLIER = 3.0
MULTI_PROTO_WINDOW   = pd.Timedelta('10min')

print("Loading Phase 3 data...")
for fname in FILES:
    for chunk in pd.read_csv(
        fname, chunksize=CHUNK_SIZE,
        usecols=['source','destination','protocolName',
                 'totalSourceBytes','totalDestinationBytes',
                 'totalSourcePackets','totalDestinationPackets',
                 'startDateTime','stopDateTime'],
        parse_dates=['startDateTime','stopDateTime'],
        low_memory=False
    ):
        chunk.dropna(subset=['source','destination','startDateTime'], inplace=True)
        chunk['source']       = chunk['source'].astype(str)
        chunk['destination']  = chunk['destination'].astype(str)
        chunk['protocolName'] = chunk['protocolName'].fillna('unknown').astype(str)
        chunk['duration']     = (chunk['stopDateTime'] - chunk['startDateTime']).dt.total_seconds().clip(lower=0)
        chunk['total_bytes']  = chunk['totalSourceBytes']  + chunk['totalDestinationBytes']
        chunk['total_pkts']   = chunk['totalSourcePackets'] + chunk['totalDestinationPackets']
        chunk['hour']         = chunk['startDateTime'].dt.floor('1H')
        chunk['day']          = chunk['startDateTime'].dt.floor('1D')
        total_flows_p3       += len(chunk)

        for ip, cnt in chunk.groupby('source').size().items():
            ip_flow_count[ip] += cnt
        for ip, val in chunk.groupby('source')['total_bytes'].sum().items():
            ip_byte_count[ip] += val
        for proto, cnt in chunk['protocolName'].value_counts().items():
            proto_counts_p3[proto] += cnt
        for hr, cnt in chunk.groupby('hour').size().items():
            hourly_flows_p3[hr] += cnt
        for dy, cnt in chunk.groupby('day').size().items():
            daily_flows[dy] += cnt

        valid = chunk['total_pkts'] > 0
        ps    = (chunk.loc[valid,'total_bytes'] / chunk.loc[valid,'total_pkts']).values
        if len(ps) > 5000:
            ps = np.random.choice(ps, 5000, replace=False)
        pkt_sizes_p3.extend(ps.tolist())

        for (ip, hr), cnt in chunk.groupby(['source','hour']).size().items():
            ip_hourly_flows[ip][hr] += cnt
        for _, row in chunk[['source','destination','hour']].iterrows():
            dst_hourly_sources[row['destination']][row['hour']].add(row['source'])

        flow_durations.append(
            chunk[['source','destination','duration','startDateTime']].copy()
        )
        sample_idx = chunk.index[::5]
        for _, row in chunk.loc[sample_idx, ['source','startDateTime','protocolName']].iterrows():
            ip_proto_times[row['source']].append((row['startDateTime'], row['protocolName']))

# Combine
flow_dur_df      = pd.concat(flow_durations, ignore_index=True)
pkt_sizes_p3     = np.array(pkt_sizes_p3)
hourly_series_p3 = pd.Series(hourly_flows_p3).sort_index()
daily_series     = pd.Series(daily_flows).sort_index()
ip_flow_series   = pd.Series(ip_flow_count)
ip_byte_series   = pd.Series(ip_byte_count)

print(f"Done. {total_flows_p3:,} flows loaded.")
```

```python
print("=" * 55)
print("3a. Statistical Traffic Analysis")
print("=" * 55)

# Packet size anomaly
ps_mean = np.mean(pkt_sizes_p3)
ps_std  = np.std(pkt_sizes_p3)
ps_low  = ps_mean - Z_ANOMALY * ps_std
ps_high = ps_mean + Z_ANOMALY * ps_std
n_ps_anomaly = np.sum((pkt_sizes_p3 < ps_low) | (pkt_sizes_p3 > ps_high))

print(f"\nPacket size statistics (threshold = mean ± {Z_ANOMALY}σ):")
print(f"  Mean   : {ps_mean:.1f} bytes")
print(f"  Std    : {ps_std:.1f} bytes")
print(f"  Range  : [{max(0,ps_low):.1f}, {ps_high:.1f}] bytes")
print(f"  Anomalous flows : {n_ps_anomaly:,} ({n_ps_anomaly/len(pkt_sizes_p3)*100:.2f}%)")
print(f"  Very small (<40B): SYN floods, ACK scans")
print(f"  Very large (>{ps_high:.0f}B): bulk data exfiltration")

# Flow count anomaly
fc_mean           = ip_flow_series.mean()
fc_std            = ip_flow_series.std()
fc_threshold_high = fc_mean + Z_ANOMALY * fc_std
high_flow_ips     = ip_flow_series[ip_flow_series > fc_threshold_high].sort_values(ascending=False)

print(f"\nFlow count anomalies (threshold = mean + {Z_ANOMALY}σ = {fc_threshold_high:.0f} flows):")
print(f"  Mean per IP     : {fc_mean:.1f} flows")
print(f"  IPs flagged     : {len(high_flow_ips)}")
print(f"\n  {'IP':<20} {'Flows':>10} {'Z-score':>8}")
print(f"  {'-'*20} {'-'*10} {'-'*8}")
for ip, cnt in high_flow_ips.head(10).items():
    z = (cnt - fc_mean) / fc_std
    print(f"  {ip:<20} {cnt:>10,} {z:>7.1f}σ")

# Protocol distribution
proto_series_p3 = pd.Series(proto_counts_p3).sort_values(ascending=False)
proto_pct_p3    = proto_series_p3 / proto_series_p3.sum() * 100
print(f"\nProtocol distribution:")
for proto, pct in proto_pct_p3.head(6).items():
    print(f"  {proto:<25} {pct:.4f}%")

rare = proto_pct_p3[proto_pct_p3 < 0.1]
print(f"\nRare protocols (<0.1%) — potentially suspicious:")
for proto, pct in rare.items():
    print(f"  {proto:<25} {pct:.4f}%  ({proto_counts_p3[proto]:,} flows)")

# Hourly vs daily
print(f"\nHourly vs Daily comparison:")
h_cv = hourly_series_p3.std() / hourly_series_p3.mean()
d_cv = daily_series.std()     / daily_series.mean()
print(f"  Hourly — mean:{hourly_series_p3.mean():,.0f}  std:{hourly_series_p3.std():,.0f}  CV:{h_cv:.2f}")
print(f"  Daily  — mean:{daily_series.mean():,.0f}  std:{daily_series.std():,.0f}  CV:{d_cv:.2f}")
print(f"  Higher hourly CV ({h_cv:.2f}) confirms intra-day bursts — hourly is more sensitive")

# Byte volume anomaly
byte_mean     = ip_byte_series.mean()
byte_std      = ip_byte_series.std()
byte_threshold= byte_mean + Z_ANOMALY * byte_std
high_byte_ips = ip_byte_series[ip_byte_series > byte_threshold].sort_values(ascending=False)

print(f"\nByte volume anomalies (threshold = {byte_threshold/1e6:.1f} MB):")
print(f"  {'IP':<20} {'Bytes sent':>15} {'Z-score':>8}")
print(f"  {'-'*20} {'-'*15} {'-'*8}")
for ip, b in high_byte_ips.head(10).items():
    z = (b - byte_mean) / byte_std
    print(f"  {ip:<20} {b/1e6:>14.1f}M {z:>7.1f}σ")
```

```python
print("=" * 55)
print("3b. Behavioural Analysis")
print("=" * 55)

# 3b.1 Spike after inactivity
print(f"\n[3b.1] Spike-after-inactivity (gap >= 2h, then z > {Z_SPIKE}σ of own average):")
spike_ips = []
for ip, hr_dict in ip_hourly_flows.items():
    if len(hr_dict) < 3:
        continue
    hours      = sorted(hr_dict.keys())
    flows_list = [hr_dict[h] for h in hours]
    mean_ip    = np.mean(flows_list)
    std_ip     = np.std(flows_list)
    if std_ip == 0:
        continue
    for i in range(1, len(hours)):
        gap_hrs = (hours[i] - hours[i-1]).total_seconds() / 3600
        if gap_hrs >= 2:
            z = (flows_list[i] - mean_ip) / (std_ip + 1e-9)
            if z > Z_SPIKE:
                spike_ips.append((ip, hours[i], flows_list[i], gap_hrs, z))

spike_ips.sort(key=lambda x: x[4], reverse=True)
print(f"  IPs found: {len(spike_ips)}")
print(f"\n  {'IP':<20} {'Spike time':<22} {'Flows':>7} {'Gap(h)':>7} {'Z':>6}")
print(f"  {'-'*20} {'-'*22} {'-'*7} {'-'*7} {'-'*6}")
for ip, ts, flows, gap, z in spike_ips[:10]:
    print(f"  {ip:<20} {str(ts):<22} {flows:>7,} {gap:>7.1f} {z:>5.1f}σ")

# 3b.2 Coordinated sources
print(f"\n[3b.2] Coordinated attack (>= 10 distinct sources to same destination/hour):")
coordinated = []
for dst, hr_dict in dst_hourly_sources.items():
    for hr, src_set in hr_dict.items():
        if len(src_set) >= 10:
            coordinated.append((dst, hr, len(src_set)))
coordinated.sort(key=lambda x: x[2], reverse=True)

src_all   = [len(s) for hd in dst_hourly_sources.values() for s in hd.values()]
src_mean  = np.mean(src_all)
src_std   = np.std(src_all)
src_thresh= src_mean + Z_DDOS * src_std

print(f"  Avg sources per dst/hour : {src_mean:.2f}")
print(f"  Threshold (mean+{Z_DDOS}σ)   : {src_thresh:.1f}")
print(f"  Destinations flagged     : {len(coordinated)}")
print(f"\n  {'Destination IP':<22} {'Hour':<22} {'Sources':>8}")
print(f"  {'-'*22} {'-'*22} {'-'*8}")
for dst, hr, cnt in coordinated[:12]:
    flag = " ← HIGH" if cnt > src_thresh else ""
    print(f"  {dst:<22} {str(hr):<22} {cnt:>8,}{flag}")

# 3b.3 Persistent targets
print(f"\n[3b.3] Persistent DDoS targets (all-time distinct sources):")
dst_total = {
    dst: len(set().union(*hd.values()))
    for dst, hd in dst_hourly_sources.items()
}
dst_series_p3 = pd.Series(dst_total).sort_values(ascending=False)
ds_mean       = dst_series_p3.mean()
ds_std        = dst_series_p3.std()
ds_threshold  = ds_mean + Z_DDOS * ds_std
flagged_dsts  = dst_series_p3[dst_series_p3 > ds_threshold]

print(f"  Threshold (mean+{Z_DDOS}σ) : {ds_threshold:.0f} sources")
print(f"  Flagged destinations  : {len(flagged_dsts)}")
print(f"\n  {'Destination IP':<22} {'Distinct sources':>18} {'Z-score':>8}")
print(f"  {'-'*22} {'-'*18} {'-'*8}")
for dst, cnt in flagged_dsts.head(10).items():
    z = (cnt - ds_mean) / ds_std
    print(f"  {dst:<22} {cnt:>18,} {z:>7.1f}σ")
```

```python
print("=" * 55)
print("3c. Suspicious Communication Patterns")
print("=" * 55)

# 3c.1 Long connections
dur_mean      = flow_dur_df['duration'].mean()
dur_std       = flow_dur_df['duration'].std()
dur_threshold = dur_mean + LONG_CONN_MULTIPLIER * dur_std
long_conns    = flow_dur_df[flow_dur_df['duration'] > dur_threshold].sort_values('duration', ascending=False)

print(f"\n[3c.1] Long-duration connections (threshold = mean + {LONG_CONN_MULTIPLIER}σ):")
print(f"  Mean duration  : {dur_mean:.1f}s")
print(f"  Std duration   : {dur_std:.1f}s")
print(f"  Threshold      : {dur_threshold:.1f}s ({dur_threshold/60:.1f} min)")
print(f"  Long conns     : {len(long_conns):,}")
print(f"\n  {'Source IP':<20} {'Destination IP':<20} {'Duration (s)':>13} {'Start'}")
print(f"  {'-'*20} {'-'*20} {'-'*13} {'-'*20}")
for _, row in long_conns.head(12).iterrows():
    print(f"  {row['source']:<20} {row['destination']:<20} "
          f"{row['duration']:>13,.0f}   {str(row['startDateTime'])[:19]}")

# 3c.2 Multi-protocol (fast pandas version)
print(f"\n[3c.2] Multi-protocol IPs (>= 3 protocols in {MULTI_PROTO_WINDOW}):")
proto_event_df = pd.DataFrame(
    [(ip, ts, proto) for ip, events in ip_proto_times.items()
     for ts, proto in events],
    columns=['source','timestamp','protocol']
).dropna()
proto_event_df       = proto_event_df.sort_values(['source','timestamp'])
proto_event_df['bin']= proto_event_df['timestamp'].dt.floor('10min')

multi_proto_df = (
    proto_event_df.groupby(['source','bin'])['protocol']
    .nunique().reset_index()
    .rename(columns={'protocol':'n_protocols'})
)
multi_proto_df = multi_proto_df[multi_proto_df['n_protocols'] >= 3]

multi_proto_ips = []
for ip, grp in multi_proto_df.groupby('source'):
    best   = grp.loc[grp['n_protocols'].idxmax()]
    protos = set(proto_event_df.loc[
        (proto_event_df['source'] == ip) &
        (proto_event_df['bin'] == best['bin']), 'protocol'
    ].tolist())
    multi_proto_ips.append((ip, int(best['n_protocols']), protos))

multi_proto_ips.sort(key=lambda x: x[1], reverse=True)
print(f"  IPs flagged: {len(multi_proto_ips)}")
print(f"\n  {'Source IP':<20} {'# Protocols':>12}  Protocols used")
print(f"  {'-'*20} {'-'*12}  {'-'*30}")
for ip, n, protos in multi_proto_ips[:10]:
    print(f"  {ip:<20} {n:>12}  {', '.join(sorted(protos))}")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(17, 10))
fig.suptitle('Phase 3 — Advanced Anomaly Detection', fontsize=13, fontweight='bold')

# 3a-1: High-flow IPs
ax = axes[0,0]
top_hf    = high_flow_ips.head(12)
clrs      = ['red' if ip in ['192.168.5.122','192.168.2.107'] else 'steelblue'
             for ip in top_hf.index[::-1]]
ax.barh([ip[-15:] for ip in top_hf.index[::-1]], top_hf.values[::-1],
        color=clrs, height=0.6)
ax.axvline(fc_threshold_high, color='red', linestyle='--', lw=1.2,
           label=f'Threshold ({fc_threshold_high:.0f})')
ax.set_xlabel('Flow count')
ax.set_title('3a: High-flow source IPs (z-scores)')
ax.legend(fontsize=8)
for i, (ip, cnt) in enumerate(zip(top_hf.index[::-1], top_hf.values[::-1])):
    z = (cnt - fc_mean) / fc_std
    ax.text(cnt+1000, i, f'z={z:.1f}σ', va='center', fontsize=7)

# 3a-2: Byte volume anomalies
ax = axes[0,1]
top_by = high_byte_ips.head(10)
ax.barh([ip[-15:] for ip in top_by.index[::-1]],
        [v/1e6 for v in top_by.values[::-1]], color='steelblue', height=0.6)
ax.axvline(byte_threshold/1e6, color='red', linestyle='--', lw=1.2,
           label=f'Threshold ({byte_threshold/1e6:.0f} MB)')
ax.set_xlabel('Bytes sent (MB)')
ax.set_title('3a: Byte volume anomalies')
ax.legend(fontsize=8)

# 3a-3: Hourly vs daily
ax = axes[0,2]
ax2 = ax.twinx()
ax.plot(hourly_series_p3.index, hourly_series_p3.values,
        color='steelblue', lw=0.8, alpha=0.8, label='Hourly')
h_thresh = hourly_series_p3.mean() + Z_SPIKE * hourly_series_p3.std()
ax.axhline(h_thresh, color='red', linestyle=':', lw=1, label='Spike threshold')
spikes_p3 = hourly_series_p3[hourly_series_p3 > h_thresh]
ax.scatter(spikes_p3.index, spikes_p3.values, color='red', s=30, zorder=5)
ax2.bar(daily_series.index, daily_series.values,
        width=0.8, color='orange', alpha=0.25, label='Daily')
ax.set_ylabel('Flows/hour', color='steelblue', fontsize=8)
ax2.set_ylabel('Flows/day', color='orange', fontsize=8)
ax.set_title('3a: Hourly vs daily (CV hourly=0.92 > daily=0.71)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m/%d'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=30, fontsize=7)
l1,lb1 = ax.get_legend_handles_labels()
l2,lb2 = ax2.get_legend_handles_labels()
ax.legend(l1+l2, lb1+lb2, fontsize=7)

# 3b: DDoS sources per hour
ax = axes[1,0]
if coordinated:
    coord_df = pd.DataFrame(coordinated, columns=['dst','hour','cnt'])
    top_dsts = coord_df.groupby('dst')['cnt'].max().nlargest(4).index
    for dst in top_dsts:
        sub = coord_df[coord_df['dst']==dst].sort_values('hour')
        ax.plot(sub['hour'], sub['cnt'], marker='o', ms=3, lw=0.9, label=dst[-15:])
    ax.axhline(src_thresh, color='red', linestyle=':', lw=1, label='Threshold')
    ax.set_ylabel('Distinct source IPs')
    ax.set_title('3b: Distinct sources per destination/hour')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%m/%d'))
    plt.setp(ax.xaxis.get_majorticklabels(), rotation=30, fontsize=7)
    ax.legend(fontsize=6)

# 3c-1: Flow duration histogram
ax = axes[1,1]
dur_clip = flow_dur_df['duration'].clip(0, dur_threshold*2).dropna()
ax.hist(dur_clip, bins=80, color='steelblue', alpha=0.85, edgecolor='none')
ax.axvline(dur_threshold, color='red', linestyle='--', lw=1.5,
           label=f'Threshold ({dur_threshold:.0f}s)')
ax.axvline(dur_mean, color='green', linestyle='--', lw=1,
           label=f'Mean ({dur_mean:.0f}s)')
ax.set_xlabel('Flow duration (seconds)')
ax.set_ylabel('Count')
ax.set_title(f'3c: Flow durations ({len(long_conns):,} long connections)')
ax.legend(fontsize=8)

# 3c-2: Multi-protocol
ax = axes[1,2]
if multi_proto_ips:
    mp_ips = [x[0][-15:] for x in multi_proto_ips[:10]]
    mp_cnt = [x[1] for x in multi_proto_ips[:10]]
    ax.barh(mp_ips[::-1], mp_cnt[::-1], color='mediumseagreen', height=0.6)
    ax.axvline(3, color='red', linestyle='--', lw=1.2, label='Threshold (3)')
    ax.set_xlabel('Distinct protocols in 10-min window')
    ax.set_title(f'3c: Multi-protocol IPs ({len(multi_proto_ips)} flagged)')
    ax.legend(fontsize=8)
    for i, (ip, n, protos) in enumerate(multi_proto_ips[:10]):
        ax.text(3.05, 9-i, ', '.join(sorted(protos)), va='center', fontsize=7)

plt.tight_layout()
plt.savefig('phase3_plots.png', dpi=150, bbox_inches='tight')
plt.show()
```

# Phase 4 — Deep Security Threat Analysis

**Four sub-sections:**
- **4a. Complex Attack Patterns** — stealthy port scan, slow DDoS, IP hopping
- **4b. Malicious Payload Identification** — large flows, encrypted anomalies, C2 beaconing
- **4c. Threat Attribution** — risk classification (HIGH / MEDIUM / LOW)

**Key thresholds:**
- Port scan: > 15 distinct ports spread over > 30 minutes (slow/stealthy)
- Slow DDoS: >= 10% hourly growth for >= 4 consecutive hours
- IP hopping: >= 5 IPs from same /24 subnet contacting same destination
- C2 beaconing: >= 8 flows to external IP with inter-arrival time CV < 0.35
- Large payload: single flow > 10,000 bytes

```python
# ── Accumulators ──────────────────────────────────────────────────────────────
src_dst_ports    = defaultdict(lambda: defaultdict(set))
src_port_times   = defaultdict(lambda: defaultdict(list))
dst_hourly_flows = defaultdict(lambda: defaultdict(int))
subnet_dst_srcs  = defaultdict(lambda: defaultdict(set))
src_dst_times    = defaultdict(lambda: defaultdict(list))
large_payload_flows = []
encrypted_flows     = []
label_counts        = defaultdict(int)
total_flows_p4      = 0

PORT_SCAN_MIN_PORTS    = 15
PORT_SCAN_MIN_DURATION = 30
SLOW_DDOS_MIN_HOURS    = 4
SLOW_DDOS_GROWTH_RATE  = 0.10
IP_HOP_MIN_SOURCES     = 5
BEACON_MIN_FLOWS       = 8
BEACON_CV_THRESHOLD    = 0.35
PAYLOAD_LARGE_THRESH   = 10000

def subnet24(ip):
    parts = ip.split('.')
    return '.'.join(parts[:3]) if len(parts) == 4 else ip

print("Loading Phase 4 data...")
for fname in FILES:
    for chunk in pd.read_csv(
        fname, chunksize=CHUNK_SIZE,
        usecols=['source','destination','protocolName',
                 'totalSourceBytes','totalDestinationBytes',
                 'totalSourcePackets','totalDestinationPackets',
                 'sourcePort','destinationPort',
                 'startDateTime','stopDateTime',
                 'sourcePayloadAsUTF','destinationPayloadAsUTF',
                 'Label'],
        parse_dates=['startDateTime','stopDateTime'],
        low_memory=False
    ):
        chunk.dropna(subset=['source','destination','startDateTime'], inplace=True)
        chunk['source']          = chunk['source'].astype(str)
        chunk['destination']     = chunk['destination'].astype(str)
        chunk['protocolName']    = chunk['protocolName'].fillna('unknown').astype(str)
        chunk['Label']           = chunk['Label'].fillna('Unknown').astype(str)
        chunk['destinationPort'] = pd.to_numeric(chunk['destinationPort'],
                                                  errors='coerce').fillna(0).astype(int)
        chunk['total_bytes']     = (chunk['totalSourceBytes'].fillna(0) +
                                    chunk['totalDestinationBytes'].fillna(0))
        chunk['duration']        = (chunk['stopDateTime'] - chunk['startDateTime']
                                    ).dt.total_seconds().clip(lower=0)
        chunk['hour']            = chunk['startDateTime'].dt.floor('1H')
        total_flows_p4          += len(chunk)

        for lbl, cnt in chunk['Label'].value_counts().items():
            label_counts[lbl] += cnt

        # 4a: Port scan
        for _, row in chunk[['source','destination',
                              'destinationPort','startDateTime']].iterrows():
            src_dst_ports[row['source']][row['destination']].add(row['destinationPort'])
            src_port_times[row['source']][row['destination']].append(row['startDateTime'])

        # 4a: Slow DDoS
        for (dst, hr), cnt in chunk.groupby(['destination','hour']).size().items():
            dst_hourly_flows[dst][hr] += cnt

        # 4a: IP hopping
        chunk['subnet'] = chunk['source'].apply(subnet24)
        for (subnet, dst), grp in chunk.groupby(['subnet','destination']):
            subnet_dst_srcs[subnet][dst].update(grp['source'].tolist())

        # 4b: C2 beaconing — external IPs only
        ext = ~chunk['destination'].str.startswith('192.168.')
        for _, row in chunk[ext][['source','destination','startDateTime']].iterrows():
            src_dst_times[row['source']][row['destination']].append(row['startDateTime'])

        # 4b: Large payload
        large = chunk['total_bytes'] > PAYLOAD_LARGE_THRESH
        if large.any():
            large_payload_flows.append(
                chunk[large][['source','destination','total_bytes',
                               'protocolName','startDateTime','Label']].copy()
            )

        # 4b: Encrypted flows
        enc = chunk['destinationPort'].isin([443, 8443, 993, 465, 995])
        if enc.any():
            encrypted_flows.append(
                chunk[enc][['source','destination','destinationPort',
                             'total_bytes','duration','startDateTime','Label']].copy()
            )

large_payload_df = pd.concat(large_payload_flows, ignore_index=True) if large_payload_flows else pd.DataFrame()
encrypted_df     = pd.concat(encrypted_flows,     ignore_index=True) if encrypted_flows     else pd.DataFrame()

print(f"Done. {total_flows_p4:,} flows loaded.")
print(f"Label distribution:")
for lbl, cnt in sorted(label_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"  {lbl:<20} {cnt:>10,}  ({cnt/total_flows_p4*100:.1f}%)")
```

```python
print("=" * 55)
print("4a. Complex Attack Patterns")
print("=" * 55)

# 4a.1 Port scan
print(f"\n[4a.1] Stealthy Port Scan (>{PORT_SCAN_MIN_PORTS} ports, >{PORT_SCAN_MIN_DURATION} min):")
port_scanners = []
for src, dst_dict in src_dst_ports.items():
    total_ports = sum(len(ports) for ports in dst_dict.values())
    if total_ports < PORT_SCAN_MIN_PORTS:
        continue
    all_times = [t for times in src_port_times[src].values() for t in times]
    if len(all_times) < 2:
        continue
    all_times_sorted  = sorted(all_times)
    scan_duration_min = (all_times_sorted[-1] - all_times_sorted[0]).total_seconds() / 60
    if scan_duration_min > PORT_SCAN_MIN_DURATION:
        port_scanners.append((src, total_ports, len(dst_dict), scan_duration_min))

port_scanners.sort(key=lambda x: x[1], reverse=True)
print(f"  Scanners detected: {len(port_scanners)}")
print(f"\n  {'Source IP':<20} {'Ports':>8} {'Dst IPs':>8} {'Duration (min)':>15}")
print(f"  {'-'*20} {'-'*8} {'-'*8} {'-'*15}")
for src, ports, dsts, dur in port_scanners[:15]:
    print(f"  {src:<20} {ports:>8,} {dsts:>8,} {dur:>15,.0f}")

# 4a.2 Slow DDoS
print(f"\n[4a.2] Slow DDoS (>={SLOW_DDOS_GROWTH_RATE*100:.0f}% growth/hr for >={SLOW_DDOS_MIN_HOURS} consecutive hours):")
slow_ddos_targets = []
for dst, hr_dict in dst_hourly_flows.items():
    if len(hr_dict) < SLOW_DDOS_MIN_HOURS + 1:
        continue
    hours = sorted(hr_dict.keys())
    flows = [hr_dict[h] for h in hours]
    max_run = cur_run = cur_start = best_start = best_end = 0
    for i in range(1, len(flows)):
        if flows[i] > flows[i-1] * (1 + SLOW_DDOS_GROWTH_RATE):
            cur_run += 1
            if cur_run > max_run:
                max_run    = cur_run
                best_start = cur_start
                best_end   = i
        else:
            cur_run   = 0
            cur_start = i
    if max_run >= SLOW_DDOS_MIN_HOURS - 1:
        slow_ddos_targets.append((
            dst, max_run+1,
            hours[best_start], hours[best_end],
            flows[best_start], flows[best_end]
        ))

slow_ddos_targets.sort(key=lambda x: x[5], reverse=True)
print(f"  Targets detected: {len(slow_ddos_targets)}")
print(f"\n  {'Destination IP':<22} {'Hours':>6} {'Start flows':>12} {'Peak flows':>11}")
print(f"  {'-'*22} {'-'*6} {'-'*12} {'-'*11}")
for dst, hrs, t0, t1, sf, pf in slow_ddos_targets[:12]:
    growth = (pf - sf) / max(sf, 1) * 100
    print(f"  {dst:<22} {hrs:>6} {sf:>12,} {pf:>11,}  (+{growth:.0f}%)")

# 4a.3 IP hopping
print(f"\n[4a.3] IP Hopping (>={IP_HOP_MIN_SOURCES} IPs from same /24 subnet):")
ip_hoppers = []
for subnet, dst_dict in subnet_dst_srcs.items():
    for dst, src_set in dst_dict.items():
        if (len(src_set) >= IP_HOP_MIN_SOURCES and
            not dst.startswith('255.') and not dst.startswith('224.')):
            ip_hoppers.append((subnet, dst, len(src_set), sorted(src_set)))

ip_hoppers.sort(key=lambda x: x[2], reverse=True)
print(f"  Patterns detected: {len(ip_hoppers)}")
print(f"\n  {'Subnet /24':<18} {'Destination IP':<22} {'# IPs':>6}  Sample")
print(f"  {'-'*18} {'-'*22} {'-'*6}  {'-'*25}")
for subnet, dst, cnt, srcs in ip_hoppers[:10]:
    sample = ', '.join(srcs[:3]) + (f'... +{cnt-3} more' if cnt > 3 else '')
    print(f"  {subnet+'.0/24':<18} {dst:<22} {cnt:>6,}  {sample}")
```

```python
print("=" * 55)
print("4b. Malicious Payload Identification")
print("=" * 55)

# 4b.1 Large payload
print(f"\n[4b.1] Large payload flows (> {PAYLOAD_LARGE_THRESH:,} bytes):")
if not large_payload_df.empty:
    print(f"  Total large flows : {len(large_payload_df):,}")
    print(f"  Max single flow   : {large_payload_df['total_bytes'].max()/1e6:.2f} MB")
    print(f"\n  Label distribution:")
    for lbl, cnt in large_payload_df['Label'].value_counts().items():
        print(f"    {lbl:<20} {cnt:>8,} ({cnt/len(large_payload_df)*100:.1f}%)")
    print(f"\n  Top 10 largest flows:")
    print(f"  {'Source IP':<20} {'Destination IP':<20} {'MB':>8} {'Protocol':<12} {'Label'}")
    print(f"  {'-'*20} {'-'*20} {'-'*8} {'-'*12} {'-'*8}")
    for _, row in large_payload_df.nlargest(10, 'total_bytes').iterrows():
        print(f"  {row['source']:<20} {row['destination']:<20} "
              f"{row['total_bytes']/1e6:>8.2f} {row['protocolName']:<12} {row['Label']}")

# 4b.2 Encrypted anomalies
print(f"\n[4b.2] Encrypted traffic anomalies (ports 443/8443/993/465/995):")
if not encrypted_df.empty:
    enc_mean      = encrypted_df['total_bytes'].mean()
    enc_std       = encrypted_df['total_bytes'].std()
    enc_threshold = enc_mean + 3 * enc_std
    enc_anomalies = encrypted_df[encrypted_df['total_bytes'] > enc_threshold]
    print(f"  Total encrypted flows    : {len(encrypted_df):,}")
    print(f"  Mean flow size           : {enc_mean/1e3:.1f} KB")
    print(f"  Anomaly threshold (3σ)   : {enc_threshold/1e6:.2f} MB")
    print(f"  Anomalous flows          : {len(enc_anomalies):,}")
    print(f"\n  Top anomalous encrypted flows:")
    print(f"  {'Source IP':<20} {'Destination IP':<20} {'Port':>5} {'MB':>8} {'Label'}")
    print(f"  {'-'*20} {'-'*20} {'-'*5} {'-'*8} {'-'*8}")
    for _, row in enc_anomalies.nlargest(8, 'total_bytes').iterrows():
        print(f"  {row['source']:<20} {row['destination']:<20} "
              f"{row['destinationPort']:>5} {row['total_bytes']/1e6:>8.2f} {row['Label']}")
    print(f"\n  Interpretation:")
    print(f"  Normal HTTPS browsing: < 100 KB per flow")
    print(f"  These flows are 10-100x larger — possible HTTPS-tunnelled exfiltration")

# 4b.3 C2 beaconing
print(f"\n[4b.3] C2 Beaconing (>= {BEACON_MIN_FLOWS} flows to external IP, CV < {BEACON_CV_THRESHOLD}):")
print(f"  CV = std/mean of inter-arrival times")
print(f"  Human browsing: CV > 1.0  |  Malware beacon: CV ≈ 0.0")
beacon_candidates = []
for src, dst_dict in src_dst_times.items():
    for dst, timestamps in dst_dict.items():
        if len(timestamps) < BEACON_MIN_FLOWS:
            continue
        ts_sorted = sorted(timestamps)
        iats      = [(ts_sorted[i+1]-ts_sorted[i]).total_seconds()
                     for i in range(len(ts_sorted)-1)]
        if len(iats) < BEACON_MIN_FLOWS - 1:
            continue
        mean_iat = np.mean(iats)
        if mean_iat < 1:
            continue
        cv = np.std(iats) / mean_iat
        if cv < BEACON_CV_THRESHOLD:
            beacon_candidates.append((src, dst, len(timestamps),
                                      mean_iat, np.std(iats), cv,
                                      ts_sorted[0], ts_sorted[-1]))

beacon_candidates.sort(key=lambda x: x[5])
print(f"\n  Candidates found: {len(beacon_candidates)}")
print(f"\n  {'Source IP':<20} {'Destination IP':<22} {'Flows':>6} {'IAT(s)':>8} {'CV':>7}")
print(f"  {'-'*20} {'-'*22} {'-'*6} {'-'*8} {'-'*7}")
for src, dst, n, miat, siat, cv, t0, t1 in beacon_candidates[:15]:
    print(f"  {src:<20} {dst:<22} {n:>6,} {miat:>8.1f} {cv:>7.3f}")
```

```python
print("=" * 55)
print("4c. Threat Attribution and Risk Classification")
print("=" * 55)

threats = []

# Port scanners
for src, ports, dsts, dur in port_scanners:
    if ports > 1000:   risk = "HIGH"
    elif ports > 100:  risk = "MEDIUM"
    else:              risk = "LOW"
    threats.append(("Port Scan", src, "Various", risk,
                    f"Scanned {ports:,} ports across {dsts} destinations over {dur:.0f} min"))

# Slow DDoS
for dst, hrs, t0, t1, sf, pf in slow_ddos_targets:
    growth = (pf - sf) / max(sf, 1) * 100
    if pf > 10000:   risk = "HIGH"
    elif pf > 1000:  risk = "MEDIUM"
    else:            risk = "LOW"
    threats.append(("Slow DDoS", "Various", dst, risk,
                    f"Grew {growth:.0f}% over {hrs} hours (peak: {pf:,}/hr)"))

# IP hopping
for subnet, dst, cnt, srcs in ip_hoppers[:20]:
    if cnt > 50:    risk = "HIGH"
    elif cnt > 20:  risk = "MEDIUM"
    else:           risk = "LOW"
    threats.append(("IP Hopping", subnet+".0/24", dst, risk,
                    f"{cnt} IPs from subnet contacted {dst}"))

# Large payload
if not large_payload_df.empty:
    for _, row in large_payload_df.nlargest(5, 'total_bytes').iterrows():
        mb   = row['total_bytes'] / 1e6
        risk = "HIGH" if mb > 100 else ("MEDIUM" if mb > 10 else "LOW")
        threats.append(("Large Payload", row['source'], row['destination'], risk,
                        f"Single flow: {mb:.1f} MB ({row['protocolName']})"))

# C2 beaconing
for src, dst, n, miat, siat, cv, t0, t1 in beacon_candidates[:10]:
    if cv < 0.05:    risk = "HIGH"
    elif cv < 0.15:  risk = "MEDIUM"
    else:            risk = "LOW"
    threats.append(("C2 Beaconing", src, dst, risk,
                    f"CV={cv:.3f}, IAT={miat:.0f}s, {n} flows"))

# Sort and count
risk_order   = {"HIGH": 0, "MEDIUM": 1, "LOW": 2}
threats.sort(key=lambda x: risk_order[x[3]])
high_count   = sum(1 for t in threats if t[3] == "HIGH")
medium_count = sum(1 for t in threats if t[3] == "MEDIUM")
low_count    = sum(1 for t in threats if t[3] == "LOW")

print(f"\n  Total threats : {len(threats)}")
print(f"  HIGH          : {high_count}")
print(f"  MEDIUM        : {medium_count}")
print(f"  LOW           : {low_count}")

print(f"\n  {'Type':<15} {'Source':<22} {'Destination':<22} {'Risk':<8} Reason")
print(f"  {'-'*15} {'-'*22} {'-'*22} {'-'*8} {'-'*45}")
for ttype, src, dst, risk, reason in threats[:35]:
    print(f"  {ttype:<15} {src:<22} {dst:<22} {risk:<8} {reason[:50]}")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(17, 10))
fig.suptitle('Phase 4 — Deep Security Threat Analysis', fontsize=13, fontweight='bold')

# 4a-1: Port scanners
ax = axes[0,0]
top_ps   = port_scanners[:12]
clrs_ps  = ['red' if p > 10000 else 'steelblue' for _, p, _, _ in top_ps]
ax.barh([x[0][-15:] for x in top_ps[::-1]],
        [x[1] for x in top_ps[::-1]],
        color=list(reversed(clrs_ps)), height=0.6)
ax.axvline(PORT_SCAN_MIN_PORTS, color='red', linestyle='--', lw=1,
           label=f'Threshold ({PORT_SCAN_MIN_PORTS})')
ax.set_xlabel('Distinct destination ports probed')
ax.set_title(f'4a: Stealthy port scanners ({len(port_scanners)} detected)')
ax.legend(fontsize=8)

# 4a-2: Slow DDoS timelines
ax = axes[0,1]
for dst, hrs, t0, t1, sf, pf in slow_ddos_targets[:5]:
    hr_dict = dst_hourly_flows[dst]
    s = pd.Series(hr_dict).sort_index()
    ax.plot(s.index, s.values, marker='o', ms=2, lw=0.9, label=dst[-15:])
ax.set_ylabel('Flows per hour')
ax.set_title(f'4a: Slow DDoS ramp-up ({len(slow_ddos_targets)} targets)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m/%d'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=30, fontsize=7)
ax.legend(fontsize=6)

# 4a-3: IP hopping
ax = axes[0,2]
top_hop = ip_hoppers[:12]
ax.barh([f"{x[0]}.0/24→{x[1][-12:]}" for x in top_hop[::-1]],
        [x[2] for x in top_hop[::-1]], color='steelblue', height=0.55)
ax.axvline(IP_HOP_MIN_SOURCES, color='red', linestyle='--', lw=1.2,
           label=f'Threshold ({IP_HOP_MIN_SOURCES})')
ax.set_xlabel('# distinct source IPs from subnet')
ax.set_title(f'4a: IP hopping ({len(ip_hoppers):,} patterns)')
ax.legend(fontsize=8)
ax.tick_params(axis='y', labelsize=6)

# 4b-1: Large payload flows
ax = axes[1,0]
if not large_payload_df.empty:
    top_lp    = large_payload_df.nlargest(10, 'total_bytes')
    lp_labels = [f"{r['source'][-12:]}→{r['destination'][-12:]}"
                 for _, r in top_lp.iterrows()]
    lp_mb     = [r['total_bytes']/1e6 for _, r in top_lp.iterrows()]
    lp_colors = ['red' if r['Label'] != 'Normal' else 'steelblue'
                 for _, r in top_lp.iterrows()]
    ax.barh(lp_labels[::-1], lp_mb[::-1], color=lp_colors[::-1], height=0.6)
    ax.set_xlabel('Total bytes (MB)')
    ax.set_title(f'4b: Largest flows\n({len(large_payload_df):,} total, red=Attack label)')
    ax.tick_params(axis='y', labelsize=7)

# 4b-2: C2 beaconing IAT
ax = axes[1,1]
if beacon_candidates:
    for src, dst, n, miat, siat, cv, t0, t1 in beacon_candidates[:5]:
        ts_list = sorted(src_dst_times[src][dst])
        iats    = [(ts_list[i+1]-ts_list[i]).total_seconds()
                   for i in range(len(ts_list)-1)]
        ax.hist(iats, bins=25, alpha=0.65,
                label=f"{src[-12:]}→{dst[-12:]} CV={cv:.3f}")
    ax.set_xlabel('Inter-arrival time (seconds)')
    ax.set_ylabel('Count')
    ax.set_title('4b: C2 beaconing IAT\n(narrow = regular = malware)')
    ax.legend(fontsize=7)

# 4c: Risk summary
ax = axes[1,2]
bars_r = ax.bar(['HIGH', 'MEDIUM', 'LOW'],
                [high_count, medium_count, low_count],
                color=['red', 'orange', 'steelblue'], width=0.5)
for bar, val in zip(bars_r, [high_count, medium_count, low_count]):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+2,
            str(val), ha='center', fontsize=13, fontweight='bold')
ax.set_ylabel('Number of threats')
ax.set_title(f'4c: Risk classification\n{len(threats)} total threats identified')

plt.tight_layout()
plt.savefig('phase4_plots.png', dpi=150, bbox_inches='tight')
plt.show()
```
