# Section 4 — Deep Security Threat Analysis

Building upon the anomalies identified earlier, this section performs a deeper threat-hunting analysis using the network traffic data.

**Four sub-sections:**
- **4a. Complex Attack Patterns** — stealthy port scan, slow DDoS, IP hopping
- **4b. Malicious Payload Identification** — large flows, encrypted anomalies, C2 beaconing
- **4c. Threat Attribution** — risk classification (HIGH / MEDIUM / LOW)

**Key thresholds:**
- Port scan: > 15 distinct ports spread over > 30 minutes (slow/stealthy)
- Slow DDoS: >= 10% hourly growth for >= 4 consecutive hours
- IP hopping: >= 5 IPs from same /24 subnet contacting same destination
- C2 beaconing: >= 8 flows to external IP with inter-arrival time CV < 0.35

```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

csv_files = sorted(glob.glob('b/*.csv'))
df = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)

df['startDateTime'] = pd.to_datetime(df['startDateTime'], errors='coerce')
df['stopDateTime'] = pd.to_datetime(df['stopDateTime'], errors='coerce')
df['destinationPort'] = pd.to_numeric(df['destinationPort'], errors='coerce').fillna(0).astype(int)
df['totalBytes'] = df['totalSourceBytes'] + df['totalDestinationBytes']
df['hour'] = df['startDateTime'].dt.floor('h')

for col in ['sourcePayloadAsBase64', 'sourcePayloadAsUTF', 'destinationPayloadAsBase64', 'destinationPayloadAsUTF']:
    if col in df.columns:
        df[col] = df[col].fillna('')
    else:
        df[col] = ''

print(f"Loaded {len(df):,} flows")
```

## (a) Detecting Complex Attack Patterns

### Stealthy Port Scan

Detecting sources that probe many distinct ports across potentially multiple destinations, spread over a long time period (> 30 minutes) to avoid detection. Aggregating across all destinations per source captures real scanner behaviour.

```python
PORT_SCAN_MIN_PORTS = 15
PORT_SCAN_MIN_DURATION = 30  # minutes

# ensure sourcePort is numeric
df['sourcePort'] = pd.to_numeric(df['sourcePort'], errors='coerce').fillna(0).astype(int)

scan_results = []

# aggregate across all destinations per source (real scanners hit many targets)
# count BOTH source and destination ports for full picture
for src, group in df.groupby('source'):
    dst_ports = set(group['destinationPort'].unique())
    src_ports = set(group['sourcePort'].unique())
    total_ports = len(dst_ports | src_ports)  # union of all ports used
    
    if total_ports < PORT_SCAN_MIN_PORTS:
        continue
    
    times = group['startDateTime'].sort_values()
    if len(times) < 2:
        continue
    scan_duration_min = (times.iloc[-1] - times.iloc[0]).total_seconds() / 60
    
    if scan_duration_min > PORT_SCAN_MIN_DURATION:
        n_destinations = group['destination'].nunique()
        scan_results.append({
            'source': src,
            'unique_ports': total_ports,
            'destinations': n_destinations,
            'duration_min': round(scan_duration_min, 0),
            'duration_hrs': round(scan_duration_min / 60, 1)
        })

scan_df = pd.DataFrame(scan_results).sort_values('unique_ports', ascending=False)
print(f"Stealthy port scan candidates (>{PORT_SCAN_MIN_PORTS} ports, >{PORT_SCAN_MIN_DURATION} min): {len(scan_df)}")
print()
print(f"{'Source':<20} {'Ports':>8} {'Dst IPs':>8} {'Duration':>15}")
print("-" * 55)
for _, r in scan_df.head(20).iterrows():
    print(f"{r['source']:<20} {r['unique_ports']:>8,} {r['destinations']:>8,} {r['duration_min']:>12,.0f} min")
```

### Slow DDoS Detection

Detecting **gradual ramp-up** patterns: destinations where hourly flow count grows by >= 10% for >= 4 consecutive hours. This captures slow DDoS that avoids sharp spikes but steadily exhausts resources.

```python
SLOW_DDOS_MIN_HOURS = 4
SLOW_DDOS_GROWTH_RATE = 0.10  # 10% hourly growth

# aggregate flows per (destination, hour)
dst_hourly = df.groupby(['destination', 'hour']).size().reset_index(name='flow_count')

slow_ddos_targets = []

for dst, group in dst_hourly.groupby('destination'):
    group = group.sort_values('hour')
    hours = group['hour'].values
    flows = group['flow_count'].values
    if len(flows) < SLOW_DDOS_MIN_HOURS + 1:
        continue
    
    # find longest consecutive run of >= 10% growth
    max_run = 0
    cur_run = 0
    cur_start = 0
    best_start = 0
    best_end = 0
    for i in range(1, len(flows)):
        if flows[i] > flows[i-1] * (1 + SLOW_DDOS_GROWTH_RATE):
            cur_run += 1
            if cur_run > max_run:
                max_run = cur_run
                best_start = cur_start
                best_end = i
        else:
            cur_run = 0
            cur_start = i
    
    if max_run >= SLOW_DDOS_MIN_HOURS - 1:
        start_flows = flows[best_start]
        peak_flows = flows[best_end]
        growth_pct = (peak_flows - start_flows) / max(start_flows, 1) * 100
        slow_ddos_targets.append({
            'destination': dst,
            'consecutive_hours': max_run + 1,
            'start_time': hours[best_start],
            'end_time': hours[best_end],
            'start_flows': int(start_flows),
            'peak_flows': int(peak_flows),
            'total_growth': f"{growth_pct:.0f}%"
        })

slow_ddos_df = pd.DataFrame(slow_ddos_targets).sort_values('peak_flows', ascending=False)
print(f"Slow DDoS targets (>={SLOW_DDOS_GROWTH_RATE*100:.0f}% growth/hr for >={SLOW_DDOS_MIN_HOURS} consecutive hours): {len(slow_ddos_df)}")
print()
if len(slow_ddos_df):
    print(slow_ddos_df.to_string(index=False))
else:
    print("No slow DDoS ramp-up patterns found with these thresholds.")
    # fallback: show destinations with sustained high-flow windows
    agg = df.groupby(['destination', 'hour']).agg(
        flow_count=('source', 'count'),
        unique_sources=('source', 'nunique'),
        avg_bytes=('totalSourceBytes', 'mean')
    ).reset_index()
    ft = agg['flow_count'].quantile(0.95)
    st = agg['unique_sources'].quantile(0.95)
    bt = agg['avg_bytes'].quantile(0.25)
    slow_ddos_fallback = agg[(agg['flow_count'] > ft) & (agg['unique_sources'] > st) & (agg['avg_bytes'] < bt)]
    print(f"\nFallback: High flow + many sources + small bytes (gradual exhaustion): {len(slow_ddos_fallback)}")
    if len(slow_ddos_fallback):
        print(slow_ddos_fallback.to_string(index=False))
```

### IP Hopping Detection

Detecting a single attacker using multiple IP addresses from the same /24 subnet to contact the same destination — a classic evasion technique. Threshold: >= 5 IPs from the same subnet contacting the same destination.

```python
IP_HOP_MIN_SOURCES = 5

def subnet24(ip):
    """Extract /24 subnet from an IP address"""
    parts = str(ip).split('.')
    return '.'.join(parts[:3]) if len(parts) == 4 else ip

# for each (subnet, destination) pair, count distinct source IPs
df['src_subnet'] = df['source'].apply(subnet24)

subnet_dst = df.groupby(['src_subnet', 'destination'])['source'].nunique().reset_index(name='unique_sources')
ip_hoppers = subnet_dst[subnet_dst['unique_sources'] >= IP_HOP_MIN_SOURCES].sort_values('unique_sources', ascending=False)

print(f"IP hopping threshold: >= {IP_HOP_MIN_SOURCES} IPs from same /24 subnet")
print(f"Suspicious (subnet, destination) pairs: {len(ip_hoppers)}")
print()
print("Top 20:")
print(f"{'Subnet':<18} {'Destination':<22} {'Unique Sources':>15}")
print("-" * 57)
for _, row in ip_hoppers.head(20).iterrows():
    print(f"{row['src_subnet']+'.0/24':<18} {row['destination']:<22} {row['unique_sources']:>15}")

# also show which specific IPs are involved for top cases
print("\nDetailed breakdown of top 5 cases:")
for _, row in ip_hoppers.head(5).iterrows():
    subnet = row['src_subnet']
    dst = row['destination']
    involved = df[(df['src_subnet'] == subnet) & (df['destination'] == dst)]['source'].unique()
    print(f"\n  {subnet}.0/24 -> {dst} ({len(involved)} IPs):")
    for ip in sorted(involved)[:10]:
        cnt = len(df[(df['source'] == ip) & (df['destination'] == dst)])
        print(f"    {ip}: {cnt:,} flows")
```

## (b) Malicious Payload Identification

Looking at payload data for suspicious patterns.

```python
df['src_b64_len'] = df['sourcePayloadAsBase64'].str.len()
df['dst_b64_len'] = df['destinationPayloadAsBase64'].str.len()

# unusually long payloads
long_threshold = df[['src_b64_len', 'dst_b64_len']].stack().quantile(0.995)
long_payloads = df[(df['src_b64_len'] > long_threshold) | (df['dst_b64_len'] > long_threshold)]

# repeated payloads
src_repeats = df['sourcePayloadAsBase64'].value_counts()
dst_repeats = df['destinationPayloadAsBase64'].value_counts()
common_src = src_repeats[(src_repeats > 10) & (src_repeats.index != '')].index
common_dst = dst_repeats[(dst_repeats > 10) & (dst_repeats.index != '')].index
repeated = df[df['sourcePayloadAsBase64'].isin(common_src) | df['destinationPayloadAsBase64'].isin(common_dst)]

# obfuscated payloads (base64 present but no UTF content)
obfuscated = df[
    ((df['sourcePayloadAsBase64'] != '') & (df['sourcePayloadAsUTF'] == '')) |
    ((df['destinationPayloadAsBase64'] != '') & (df['destinationPayloadAsUTF'] == ''))
]

print(f"Unusually long payloads (>99.5th pctl): {len(long_payloads):,}")
print(f"Repeated payloads (>10 occurrences): {len(repeated):,}")
print(f"Obfuscated Base64 (no UTF content): {len(obfuscated):,}")
```

### Suspicious encrypted traffic

Flows on encrypted ports (443, 22, 993) with near-empty payloads.

```python
enc_flows = df[df['destinationPort'].isin([443, 22, 993])].copy()
susp_enc = enc_flows[(enc_flows['src_b64_len'] < 50) & (enc_flows['dst_b64_len'] < 50)]

print(f"Total flows on encrypted ports: {len(enc_flows):,}")
print(f"Suspicious (tiny/no payload): {len(susp_enc):,}")
```

### C2 Beaconing Detection

Malware beacons contact C2 servers at **regular intervals**. The key signal is low coefficient of variation (CV) of inter-arrival times — perfectly periodic beacons have CV near 0. We look for source-destination pairs with >= 8 flows, CV < 0.35, and connections to external IPs.

```python
BEACON_MIN_FLOWS = 8
BEACON_CV_THRESHOLD = 0.35

df['payload_size'] = df['src_b64_len'] + df['dst_b64_len']

beacon_candidates = []

for (src, dst), group in df.groupby(['source', 'destination']):
    if len(group) < BEACON_MIN_FLOWS:
        continue
    times = group['startDateTime'].sort_values().values.astype('datetime64[s]')
    intervals = np.diff(times.astype(np.int64))
    if len(intervals) < 2:
        continue
    
    mean_iat = intervals.mean()
    std_iat = intervals.std()
    if mean_iat == 0:
        continue
    cv = std_iat / mean_iat  # coefficient of variation
    
    if cv < BEACON_CV_THRESHOLD:
        dur_hrs = float((times[-1] - times[0]).astype('timedelta64[h]').astype(float))
        beacon_candidates.append({
            'source': src,
            'destination': dst,
            'flow_count': len(group),
            'mean_iat_s': round(mean_iat, 1),
            'std_iat_s': round(std_iat, 1),
            'cv': round(cv, 4),
            'duration_hrs': round(dur_hrs, 1),
            'avg_payload': round(group['payload_size'].mean(), 1)
        })

c2_df = pd.DataFrame(beacon_candidates).sort_values('cv')
print(f"C2-like beaconing pairs (CV < {BEACON_CV_THRESHOLD}, >= {BEACON_MIN_FLOWS} flows): {len(c2_df)}")
print()
if len(c2_df):
    print(f"{'Source':<18} {'Destination':<18} {'Flows':>6} {'IAT(s)':>8} {'CV':>7} {'Hrs':>5} {'Payload':>8}")
    print("-" * 75)
    for _, r in c2_df.head(15).iterrows():
        print(f"{r['source']:<18} {r['destination']:<18} {r['flow_count']:>6} {r['mean_iat_s']:>8.1f} {r['cv']:>7.4f} {r['duration_hrs']:>5.1f} {r['avg_payload']:>8.1f}")
    print()
    print("Interpretation:")
    print("  CV near 0 = perfectly periodic (strongest C2 signal)")
    print("  CV < 0.05: HIGH risk — machine-generated timing")
    print("  CV 0.05-0.15: MEDIUM risk — likely automated with minor jitter")
    print("  CV 0.15-0.35: LOW risk — suspicious but could be legitimate polling")
```

## (c) Threat Attribution and Risk Analysis

Categorising each detected security event into risk levels with justification.

```python
risk_events = []

# HIGH risk - stealthy port scans (>24 hours = very stealthy)
for _, r in scan_df[scan_df['duration_hrs'] > 24].iterrows():
    risk_events.append({
        'risk': 'High', 'type': 'Stealthy Port Scan',
        'source': r['source'], 'destination': f"{r['destinations']} targets",
        'justification': f"{r['unique_ports']} ports across {r['destinations']} destinations over {r['duration_hrs']}h - extended reconnaissance"
    })

# HIGH risk - slow DDoS ramp-up
if len(slow_ddos_df):
    for _, r in slow_ddos_df.iterrows():
        risk_events.append({
            'risk': 'High', 'type': 'Slow DDoS (ramp-up)',
            'source': 'Multiple', 'destination': r['destination'],
            'justification': f"{r['consecutive_hours']}h consecutive growth, {r['start_flows']}->{r['peak_flows']} flows/hr ({r['total_growth']} increase)"
        })

# HIGH risk - C2 beaconing (CV < 0.05)
if len(c2_df):
    for _, r in c2_df[c2_df['cv'] < 0.05].iterrows():
        risk_events.append({
            'risk': 'High', 'type': 'C2 Beaconing',
            'source': r['source'], 'destination': r['destination'],
            'justification': f"CV={r['cv']:.4f}, IAT={r['mean_iat_s']}s, {r['flow_count']} flows over {r['duration_hrs']}h — machine-generated timing"
        })

# MEDIUM risk - shorter port scans
for _, r in scan_df[scan_df['duration_hrs'] <= 24].iterrows():
    risk_events.append({
        'risk': 'Medium', 'type': 'Port Scan (short)',
        'source': r['source'], 'destination': f"{r['destinations']} targets",
        'justification': f"{r['unique_ports']} ports over {r['duration_hrs']}h - possible recon"
    })

# MEDIUM risk - IP hopping (top cases from subnet analysis)
for _, r in ip_hoppers.head(10).iterrows():
    risk_events.append({
        'risk': 'Medium', 'type': 'IP Hopping (subnet)',
        'source': r['src_subnet'] + '.0/24', 'destination': r['destination'],
        'justification': f"{r['unique_sources']} IPs from same /24 subnet contacting same target — evasion technique"
    })

# MEDIUM risk - C2 beaconing (CV 0.05-0.15)
if len(c2_df):
    for _, r in c2_df[(c2_df['cv'] >= 0.05) & (c2_df['cv'] < 0.15)].iterrows():
        risk_events.append({
            'risk': 'Medium', 'type': 'C2 Beaconing (moderate)',
            'source': r['source'], 'destination': r['destination'],
            'justification': f"CV={r['cv']:.4f}, IAT={r['mean_iat_s']}s, {r['flow_count']} flows — automated with minor jitter"
        })

# MEDIUM risk - suspicious encrypted
risk_events.append({
    'risk': 'Medium', 'type': 'Suspicious Encrypted Traffic',
    'source': 'Various', 'destination': 'Ports 22/443/993',
    'justification': f"{len(susp_enc):,} flows with near-empty payloads on encrypted ports"
})

# MEDIUM risk - obfuscated
risk_events.append({
    'risk': 'Medium', 'type': 'Obfuscated Payloads',
    'source': 'Various', 'destination': 'Various',
    'justification': f"{len(obfuscated):,} flows with Base64 but no readable UTF content"
})

# LOW risk - C2 beaconing (CV 0.15-0.35)
if len(c2_df):
    for _, r in c2_df[(c2_df['cv'] >= 0.15) & (c2_df['cv'] < 0.35)].head(5).iterrows():
        risk_events.append({
            'risk': 'Low', 'type': 'Suspicious Periodic Traffic',
            'source': r['source'], 'destination': r['destination'],
            'justification': f"CV={r['cv']:.4f}, IAT={r['mean_iat_s']}s — could be legitimate polling"
        })

# LOW risk
risk_events.append({
    'risk': 'Low', 'type': 'Repeated Payloads',
    'source': 'Various', 'destination': 'Various',
    'justification': f"{len(repeated):,} flows with identical payloads (>10 times) - could be keep-alives"
})

risk_events.append({
    'risk': 'Low', 'type': 'Large Payloads',
    'source': 'Various', 'destination': 'Various',
    'justification': f"{len(long_payloads):,} flows above 99.5th percentile - possible data exfiltration"
})

risk_df = pd.DataFrame(risk_events)
counts = risk_df['risk'].value_counts().reindex(['High', 'Medium', 'Low'], fill_value=0)

print("Risk Summary:")
for level, cnt in counts.items():
    print(f"  {level}: {cnt} events")
print(f"  Total: {len(risk_df)} events")
```

---
### Final Risk Report and Recommendations

```python
print("=" * 70)
print("           NETWORK SECURITY RISK REPORT")
print("=" * 70)

for level in ['High', 'Medium', 'Low']:
    subset = risk_df[risk_df['risk'] == level]
    if subset.empty:
        continue
    print(f"\n[{level.upper()} RISK] - {len(subset)} event(s)")
    print("-" * 70)
    for _, row in subset.head(15).iterrows():
        print(f"  Type   : {row['type']}")
        print(f"  Source : {row['source']}")
        print(f"  Target : {row['destination']}")
        print(f"  Reason : {row['justification']}")
        print()
    if len(subset) > 15:
        print(f"  ... and {len(subset) - 15} more {level}-risk events")
        print()

print("=" * 70)
print("KEY SUSPICIOUS IPs")
print("=" * 70)

# Identify top scanner
if len(scan_df):
    top_scanner = scan_df.iloc[0]
    print(f"\n  {top_scanner['source']} — Primary scanner + C2 node")
    print(f"    - Scanned {top_scanner['unique_ports']:,} ports across {top_scanner['destinations']:,} destinations over {top_scanner['duration_hrs']}h")

# Identify top C2
if len(c2_df):
    top_c2 = c2_df.iloc[0]
    print(f"    - Beaconing every {top_c2['mean_iat_s']:.0f}s to {top_c2['destination']} (CV = {top_c2['cv']:.3f})")

# Identify top DDoS target
dst_flow_counts = df.groupby('destination').size().sort_values(ascending=False)
top_dst = dst_flow_counts.index[0]
top_dst_bytes = df[df['destination'] == top_dst]['totalDestinationBytes'].sum()
print(f"\n  {top_dst} — DDoS target")
print(f"    - {dst_flow_counts.iloc[0]:,} flows, received {top_dst_bytes/1e9:.2f} GB")

# Identify top data exfiltration suspect
src_bytes = df.groupby('source')['totalSourceBytes'].sum().sort_values(ascending=False)
top_exfil = src_bytes.index[0]
print(f"\n  {top_exfil} — Data exfiltration suspect")
print(f"    - {src_bytes.iloc[0]/1e6:,.1f} MB sent")

print(f"\n{'=' * 70}")
print("SUMMARY")
print("=" * 70)
for level, cnt in counts.items():
    print(f"  {level}: {cnt} events")
print(f"  Total: {len(risk_df)}")

print(f"\nRECOMMENDATIONS:")
if len(scan_df):
    print(f"  1. Isolate {scan_df.iloc[0]['source']} immediately — confirmed scanner + C2 activity")
print(f"  2. Block all {len(c2_df)} beaconing destinations at perimeter firewall")
print(f"  3. Rate-limit any internal IP exceeding 30,000 flows/hour")
print(f"  4. Investigate large encrypted flows on ports 443/22/993")
print(f"  5. Audit 192.168.2.0/24 subnet — multiple hosts involved in coordinated activity")
print("=" * 70)
```
