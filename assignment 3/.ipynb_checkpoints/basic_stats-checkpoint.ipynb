# Basic Statistics — Network Traffic Analysis

This notebook computes fundamental statistics over the PCAP-derived network flow dataset stored in `../b/network_analysis_data*.csv`. Each row represents a single network flow captured in 2010.

**Sections**
1. Load data
2. Total number of flows
3. Top-k most used protocols
4. Top 10 most active source and destination IPs
5. Average packet size
6. Top 3 most frequent source–destination pairs
7. IPs consistently communicating across ≥80 % of hourly windows
8. Traffic spikes (mean + 2 × std on hourly volume)
9. Variance of totalBytes per flow

```python
# ── Imports ────────────────────────────────────────────────────────────────
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (12, 5)
```

## 1. Load Data

We read all CSV files matching `../b/network_analysis_data*.csv` and concatenate them into a single DataFrame. Only the columns we actually need are loaded to keep memory usage manageable across ~2 million rows.

```python
# Columns to load (subset keeps RAM usage low)
COLS = [
    'source', 'destination', 'protocolName', 'sourcePort', 'destinationPort',
    'totalSourceBytes', 'totalDestinationBytes',
    'totalSourcePackets', 'totalDestinationPackets',
    'startDateTime', 'stopDateTime', 'Label'
]

csv_files = sorted(glob.glob('../b/*.csv'))
print(f'Found {len(csv_files)} CSV files:')
for f in csv_files:
    print(' ', f)

df = pd.concat(
    [pd.read_csv(f, usecols=lambda c: c in COLS, low_memory=False) for f in csv_files],
    ignore_index=True
)

# Parse datetimes
df['startDateTime'] = pd.to_datetime(df['startDateTime'], errors='coerce')
df['stopDateTime']  = pd.to_datetime(df['stopDateTime'],  errors='coerce')

# Ensure numeric byte/packet columns
for col in ['totalSourceBytes', 'totalDestinationBytes',
            'totalSourcePackets', 'totalDestinationPackets']:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# Derived columns used throughout the notebook
df['totalBytes']   = df['totalSourceBytes']   + df['totalDestinationBytes']
df['totalPackets'] = df['totalSourcePackets'] + df['totalDestinationPackets']

print(f'\nTotal rows loaded: {len(df):,}')
df.head(3)
```

## 2. Total Number of Flows

```python
total_flows = len(df)
print(f'Total number of network flows: {total_flows:,}')
```

## 3. Top-k Most Used Protocols

We count occurrences of each `protocolName` and display the top `k` protocols. `k` is configurable.

```python
K = 5  # configurable

top_protocols = df['protocolName'].value_counts().head(K)
print(f'Top {K} protocols by flow count:')
print(top_protocols.to_string())

fig, ax = plt.subplots()
top_protocols.plot(kind='bar', ax=ax, color=sns.color_palette('tab10', K))
ax.set_title(f'Top {K} Most Used Protocols')
ax.set_xlabel('Protocol')
ax.set_ylabel('Flow Count')
ax.tick_params(axis='x', rotation=30)
plt.tight_layout()
plt.show()
```

## 4. Top 10 Most Active Source and Destination IPs

```python
top_src = df['source'].value_counts().head(10)
top_dst = df['destination'].value_counts().head(10)

print('Top 10 Source IPs:')
print(top_src.to_string())
print('\nTop 10 Destination IPs:')
print(top_dst.to_string())

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

top_src.plot(kind='barh', ax=axes[0], color='steelblue')
axes[0].set_title('Top 10 Source IPs')
axes[0].set_xlabel('Flow Count')
axes[0].invert_yaxis()

top_dst.plot(kind='barh', ax=axes[1], color='coral')
axes[1].set_title('Top 10 Destination IPs')
axes[1].set_xlabel('Flow Count')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

## 5. Average Packet Size

Average packet size (bytes) = (totalSourceBytes + totalDestinationBytes) / (totalSourcePackets + totalDestinationPackets) per flow, then averaged across all flows.

```python
# Compute per-flow average packet size; drop flows with 0 packets to avoid division by zero
valid = df[df['totalPackets'] > 0].copy()
valid['avg_pkt_size'] = valid['totalBytes'] / valid['totalPackets']

global_avg_pkt_size = valid['avg_pkt_size'].mean()
print(f'Global average packet size: {global_avg_pkt_size:.2f} bytes')

fig, ax = plt.subplots()
# Use log-scale bins for skewed distribution
ax.hist(valid['avg_pkt_size'].clip(upper=valid['avg_pkt_size'].quantile(0.99)),
        bins=100, color='teal', edgecolor='none')
ax.axvline(global_avg_pkt_size, color='red', linestyle='--', label=f'Mean = {global_avg_pkt_size:.1f} B')
ax.set_title('Distribution of Per-Flow Average Packet Size (clipped at 99th pct)')
ax.set_xlabel('Bytes')
ax.set_ylabel('Flow Count')
ax.legend()
plt.tight_layout()
plt.show()
```

## 6. Top 3 Most Frequent Source–Destination Pairs

```python
pair_counts = (
    df.groupby(['source', 'destination'])
      .size()
      .reset_index(name='flow_count')
      .sort_values('flow_count', ascending=False)
      .head(3)
)

print('Top 3 most frequent source–destination pairs:')
print(pair_counts.to_string(index=False))
```

## 7. IPs Consistently Communicating Across ≥80 % of Hourly Windows

We bucket flows into hourly windows by `startDateTime`, then count how many distinct windows exist. An IP (source) is "consistent" if it appears in at least 80 % of those windows.

```python
CONSISTENCY_THRESHOLD = 0.80

df['hour_window'] = df['startDateTime'].dt.floor('H')

total_windows = df['hour_window'].nunique()
print(f'Total distinct hourly windows: {total_windows}')

# Count unique hours per source IP
src_windows = (
    df.groupby('source')['hour_window']
      .nunique()
      .reset_index(name='active_windows')
)
src_windows['coverage'] = src_windows['active_windows'] / total_windows

consistent_ips = src_windows[src_windows['coverage'] >= CONSISTENCY_THRESHOLD].sort_values(
    'coverage', ascending=False
)

print(f'\nIPs active in ≥{CONSISTENCY_THRESHOLD*100:.0f}% of hourly windows: {len(consistent_ips)}')
print(consistent_ips.head(20).to_string(index=False))
```

## 8. Traffic Spikes

We aggregate traffic volume (MB) by hour and flag hours where volume exceeds `mean + 2 × std` as spikes.

```python
hourly_mb = (
    df.groupby('hour_window')['totalBytes']
      .sum()
      .reset_index(name='total_bytes')
      .sort_values('hour_window')
)
hourly_mb['total_MB'] = hourly_mb['total_bytes'] / 1e6

mean_mb = hourly_mb['total_MB'].mean()
std_mb  = hourly_mb['total_MB'].std()
spike_threshold = mean_mb + 2 * std_mb

hourly_mb['is_spike'] = hourly_mb['total_MB'] > spike_threshold
spikes = hourly_mb[hourly_mb['is_spike']]

print(f'Mean hourly traffic: {mean_mb:.2f} MB')
print(f'Std:                 {std_mb:.2f} MB')
print(f'Spike threshold:     {spike_threshold:.2f} MB')
print(f'Number of spike hours: {len(spikes)}')
print(spikes[['hour_window', 'total_MB']].to_string(index=False))

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(hourly_mb['hour_window'], hourly_mb['total_MB'], lw=0.8, label='Hourly Traffic (MB)')
ax.axhline(spike_threshold, color='red', linestyle='--', label=f'Spike threshold ({spike_threshold:.1f} MB)')
ax.scatter(spikes['hour_window'], spikes['total_MB'], color='red', s=20, zorder=5, label='Spike')
ax.set_title('Hourly Traffic Volume with Spike Detection')
ax.set_xlabel('Time')
ax.set_ylabel('MB')
ax.legend()
plt.tight_layout()
plt.show()
```

## 9. Variance of totalBytes per Flow

```python
var_bytes = df['totalBytes'].var()
std_bytes = df['totalBytes'].std()
print(f'Variance of totalBytes per flow: {var_bytes:,.2f}')
print(f'Std Dev  of totalBytes per flow: {std_bytes:,.2f}')

# Distribution (log scale to handle extreme skew)
fig, ax = plt.subplots()
positive = df[df['totalBytes'] > 0]['totalBytes']
ax.hist(np.log10(positive + 1), bins=100, color='slateblue', edgecolor='none')
ax.set_title('Distribution of log10(totalBytes) per Flow')
ax.set_xlabel('log10(Bytes + 1)')
ax.set_ylabel('Flow Count')
plt.tight_layout()
plt.show()

# Summary stats
df['totalBytes'].describe().to_frame().T
```
